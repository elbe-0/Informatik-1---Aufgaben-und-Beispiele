## Aufgabe

Der Drehzahlsensor liefert den Wert `1500` (Umdrehungen pro Minute). Ein Korrekturfaktor von `2.0` soll die Drehzahl anpassen.

**(a)** Legen Sie eine Variable `drehzahl` mit dem Wert `1500` an. Geben Sie den Wert und den Datentyp mit `type()` aus.

**(b)** Dividieren Sie `drehzahl` durch den Korrekturfaktor `2.0` und speichern Sie das Ergebnis wieder in `drehzahl`. Geben Sie erneut Wert und Datentyp aus.

Was hat sich geändert? Warum?

**(c)** Testen Sie: Was passiert, wenn Sie stattdessen durch `2` (ohne Dezimalpunkt) dividieren? Ändert sich der Datentyp trotzdem?

**(d)** Wie können Sie sicherstellen, dass Ihr Ergebnis den gewünschten Datentyp hat? Schauen Sie bei den mathematischen Operatoren im Skript nach einem geeigneten Hilfsmittel.

**(e)** Testen Sie nun die Multiplikation statt der Division. Verhält sich der Datentyp genauso? Diskutieren Sie mit Ihrem Nachbarn.

```python
# === Aufgabe (a): Variable anlegen und Typ prüfen ===
# Der Sensor liefert eine Ganzzahl (int).
drehzahl = 1500
print("Drehzahl:", drehzahl)
print("Typ:", type(drehzahl))  # <class 'int'>
```

```python
# === Aufgabe (b): Division durch Korrekturfaktor 2.0 ===
# Wenn ein int durch einen float dividiert wird,
# wird das Ergebnis automatisch zu float.
drehzahl = 1500
drehzahl = drehzahl / 2.0
print("Drehzahl:", drehzahl)   # 750.0
print("Typ:", type(drehzahl))  # <class 'float'>

# Erkenntnis: Der Typ hat sich von int zu float geändert!
# Python "promoted" automatisch zu float, wenn mindestens
# ein Operand ein float ist.
```

```python
# === Aufgabe (c): Division durch 2 (Ganzzahl) ===
# Auch bei Division durch eine Ganzzahl wird das
# Ergebnis in Python immer float!
drehzahl = 1500
drehzahl = drehzahl / 2
print("Drehzahl:", drehzahl)   # 750.0
print("Typ:", type(drehzahl))  # <class 'float'>

# Erkenntnis: Der /-Operator liefert in Python IMMER
# einen float - egal ob die Operanden int oder float sind.
# Für Ganzzahl-Division gibt es den //-Operator:
drehzahl = 1500
drehzahl = drehzahl // 2
print("Drehzahl (Ganzzahl-Division):", drehzahl)  # 750
print("Typ:", type(drehzahl))  # <class 'int'>
```

```python
# === Aufgabe (d): Gewünschten Datentyp sicherstellen ===
# Mit dem Ganzzahl-Divisionsoperator // bleibt das Ergebnis int:
drehzahl = 1500
drehzahl = drehzahl // 2
print("Ganzzahl-Division:", drehzahl)
print("Typ:", type(drehzahl))  # <class 'int'>

# Alternativ: Explizite Umwandlung mit int()
drehzahl = int(1500 / 2)
print("Mit int():", drehzahl)
print("Typ:", type(drehzahl))  # <class 'int'>
```

### Aufgabe (e): Warum verhält sich Multiplikation anders?

**Kurzgefasst:** Mathematische Vorhersagbarkeit.

Das Produkt zweier Ganzzahlen ist **immer** wieder eine Ganzzahl (`5 * 2 = 10`). Es gibt keinen Fall, in dem `int * int` Nachkommastellen erzeugt. Daher gibt es für Python keinen Grund, den Datentyp zu ändern – `int` bleibt `int`.

Bei der Division hingegen kann die Rechnung einen Rest ergeben (`5 / 2 = 2.5`). Deshalb entscheidet sich Python beim `/`-Operator immer für `float` als "sicheren" Datentyp.

**Zusätzlicher Grund:** Integers haben in Python eine **unbegrenzte Genauigkeit** – sie können hunderte Stellen lang sein, ohne dass Ziffern verloren gehen. Floats sind auf 64 Bit begrenzt. Eine automatische Umwandlung bei der Multiplikation würde also bei sehr großen Zahlen zu Genauigkeitsverlust führen.
