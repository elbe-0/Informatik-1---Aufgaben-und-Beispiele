## Briefing

Das QR-Code-System funktioniert, aber Dr. Voss hat ein Problem entdeckt: Bei der letzten Charge wurden 500 Etiketten mit falschen Daten gedruckt, weil niemand die berechneten Werte vor dem Druck kontrolliert hat. Ab sofort soll das Programm eine Prüfpause einlegen, damit der Techniker die Daten kontrollieren kann, bevor der QR-Code generiert wird.

## Stakeholder-Dialog

*Dr. Lena Voss schaut genervt auf einen Stapel falscher Etiketten:*

> *"Erst prüfen, dann drucken. Ich hab keine Lust, nochmal 500 falsche Etiketten wegzuschmeißen. Das Programm soll die Daten erst als Klartext anzeigen, drei Sekunden warten und DANN erst den QR-Code erzeugen."*

## Aufgabe

Erweitern Sie das Programm aus Aufgabe 2.6:

**(a)** Fragen Sie wie zuvor Maschinenname, ID und Leistung ab. Berechnen Sie den Tagesverbrauch wie in Aufgabe 2.6.

**(b)** Geben Sie **vor** dem QR-Code alle berechneten Werte als Klartext aus, z.B.:

```
=== Datenprüfung ===
Name: Presswerk Alpha
ID: 1042
Leistung: 220.0 kW
Tagesverbrauch: 3520 kWh

QR-Code wird in 3 Sekunden generiert...
```

**(c)** Warten Sie mit `time.sleep(3)` drei Sekunden. Geben Sie danach eine Bestätigung aus und erzeugen Sie den QR-Code.

````python
import qrcode
import math
import time

# === Aufgabe (a): Daten eingeben und berechnen ===
# In der Aufgabe werden die Daten per input() abgefragt.
# Hier verwenden wir Beispielwerte für die Demonstration:
name = "Presswerk Alpha"
maschinen_id = 1042
leistung = 220.0

betriebsstunden = 16
tagesverbrauch = math.ceil(leistung * betriebsstunden)

# === Aufgabe (b): Daten zur Prüfung anzeigen ===
print()
print("=== Datenprüfung ===")
print(f"Name: {name}")
print(f"ID: {maschinen_id}")
print(f"Leistung: {leistung} kW")
print(f"Tagesverbrauch: {tagesverbrauch} kWh")
print()
print("QR-Code wird in 3 Sekunden generiert...")

# === Aufgabe (c): Prüfpause und QR-Code ===
# time.sleep(3) hält das Programm 3 Sekunden an.
# In dieser Zeit kann der Techniker die Daten prüfen.
time.sleep(3)

print("QR-Code wird generiert...")

etikett_text = f"""MotoTec Maschinenetikett
Name: {name}
ID: {maschinen_id}
Leistung: {leistung} kW
Tagesverbrauch: {tagesverbrauch} kWh"""

bild = qrcode.make(etikett_text)
display(bild)
````

## Debriefing

*Dr. Lena Voss beobachtet den Ablauf: Daten erscheinen, kurze Pause, dann der QR-Code.*

> *"So ist es richtig. Erst lesen, dann drucken. Die drei Sekunden geben dem Techniker Zeit, die Daten zu kontrollieren. Lieber drei Sekunden warten als 500 falsche Etiketten produzieren."*
