## Briefing

Die QR-Code-Etiketten funktionieren und die Prüfpause verhindert Fehler. Aber Dr. Voss will noch eine Verbesserung: Jedes Etikett soll einen Zeitstempel enthalten. Beim Scannen soll sofort ersichtlich sein, wann das Etikett erstellt wurde - damit niemand mit veralteten Daten arbeitet.

## Stakeholder-Dialog

*Dr. Lena Voss hält zwei identisch aussehende Etiketten nebeneinander:*

> *"Ich will auf jedem Etikett sehen, WANN es erstellt wurde. Sonst weiß keiner, ob die Daten noch aktuell sind. Gestern hatten wir Etiketten vom letzten Monat im Umlauf - das darf nicht passieren."*

## Aufgabe

Erweitern Sie das Programm aus Aufgabe 2.7 um einen Zeitstempel:

**(a)** Erzeugen Sie einen Zeitstempel im Format `TT.MM.JJJJ HH:MM`, z.B. `03.03.2026 14:30`.

**(b)** Fügen Sie den Zeitstempel zur Datenprüfung und zum QR-Code-Inhalt hinzu. Der QR-Code soll jetzt enthalten:

```
MotoTec Maschinenetikett
Name: Presswerk Alpha
ID: 1042
Leistung: 220.0 kW
Tagesverbrauch: 3520 kWh
Erstellt: 03.03.2026 14:30
```

````python
import qrcode
import math
import time

# In der Aufgabe werden die Daten per input() abgefragt.
# Hier verwenden wir Beispielwerte für die Demonstration:
name = "Presswerk Alpha"
maschinen_id = 1042
leistung = 220.0

# Tagesverbrauch berechnen
betriebsstunden = 16
tagesverbrauch = math.ceil(leistung * betriebsstunden)

# === Aufgabe (a): Zeitstempel erzeugen ===
# time.localtime() liefert die aktuelle Ortszeit.
# time.strftime() formatiert sie als lesbaren String.
# %d = Tag, %m = Monat, %Y = Jahr, %H = Stunde, %M = Minute
zeitstempel = time.strftime("%d.%m.%Y %H:%M", time.localtime())

# === Aufgabe (b): Datenprüfung mit Zeitstempel ===
print()
print("=== Datenprüfung ===")
print(f"Name: {name}")
print(f"ID: {maschinen_id}")
print(f"Leistung: {leistung} kW")
print(f"Tagesverbrauch: {tagesverbrauch} kWh")
print(f"Erstellt: {zeitstempel}")
print()
print("QR-Code wird in 3 Sekunden generiert...")

time.sleep(3)

# QR-Code mit Zeitstempel erzeugen
etikett_text = f"""MotoTec Maschinenetikett
Name: {name}
ID: {maschinen_id}
Leistung: {leistung} kW
Tagesverbrauch: {tagesverbrauch} kWh
Erstellt: {zeitstempel}"""

print("QR-Code wird generiert...")
bild = qrcode.make(etikett_text)
display(bild)
````
