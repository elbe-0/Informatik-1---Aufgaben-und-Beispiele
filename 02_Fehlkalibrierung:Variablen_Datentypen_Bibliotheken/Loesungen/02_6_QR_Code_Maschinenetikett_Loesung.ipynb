## Aufgabe

Schreiben Sie ein Programm, das ein QR-Code-Etikett für eine Maschine generiert:

**(a)** Fragen Sie per `input()` folgende Daten ab:
- Maschinenname (Text)
- Maschinen-ID (Ganzzahl)
- Leistung in kW (Gleitkommazahl)

**(b)** Berechnen Sie den **Tagesverbrauch** in kWh:
- Betriebsstunden pro Tag: `16`
- Berechnen Sie aus Leistung und Betriebsstunden den Tagesverbrauch
- Runden Sie den Tagesverbrauch auf die nächste Ganzzahl auf

**(c)** Erstellen Sie einen f-String mit allen Maschinendaten und generieren Sie daraus einen QR-Code. Der QR-Code soll folgenden Text enthalten:

```
MotoTec Maschinenetikett
Name: [Name]
ID: [ID]
Leistung: [Leistung] kW
Tagesverbrauch: [Verbrauch] kWh
```

Scannen Sie den QR-Code mit Ihrem Handy, um das Ergebnis zu prüfen.

> **Hinweis:** Manche QR-Scanner-Apps starten automatisch eine Websuche statt den Text anzuzeigen.

````python
import qrcode
import math

# === Aufgabe (a): Daten eingeben ===
# In der Aufgabe werden die Daten per input() abgefragt.
# Hier verwenden wir Beispielwerte für die Demonstration:
name = "Presswerk Alpha"
maschinen_id = 1042
leistung = 220.0

# === Aufgabe (b): Tagesverbrauch berechnen ===
# 16 Betriebsstunden pro Tag
betriebsstunden = 16
tagesverbrauch = leistung * betriebsstunden

# math.ceil() rundet auf die nächste Ganzzahl auf
tagesverbrauch = math.ceil(tagesverbrauch)

# === Aufgabe (c): QR-Code erzeugen ===
# f-String mit allen Daten
etikett_text = f"""MotoTec Maschinenetikett
Name: {name}
ID: {maschinen_id}
Leistung: {leistung} kW
Tagesverbrauch: {tagesverbrauch} kWh"""

print(etikett_text)
print()

# QR-Code generieren und anzeigen
bild = qrcode.make(etikett_text)
display(bild)
````
