## Briefing

MotoTec legt jeden Monat einen festen Betrag für die Anschaffung neuer Produktionsmaschinen zurück.
Diese sogenannte **Maschinenrücklage** wird jährlich mit 7 % Rendite verzinst.
Die Finanzchefin Aylin Demir möchte wissen, wie sich das Kapital über einen Zeitraum von 50 Jahren entwickelt.
Dabei soll die monatliche Sparrate vorab geprüft werden, damit nur zulässige Beträge akzeptiert werden.

## Stakeholder-Dialog

> Die monatliche Rate muss zwischen 50 und 300 Euro liegen – in 50er-Schritten.
> Alles andere akzeptiere ich nicht.“

## Beispiele

### Iteration mit `for` und `range()`

Mit `for` und `range()` können Sie eine Aktion eine bestimmte Anzahl von Malen wiederholen.
`range(1, n + 1)` erzeugt die Zahlen 1 bis n.

````python
# Beispiel: Summe der Zahlen 1 bis 5 berechnen
summe = 0
for i in range(1, 6):
    summe = summe + i
    print(f"Nach Durchlauf {i}: Summe = {summe}")
````

````python
# Beispiel: Jährliche Verzinsung eines Startkapitals
kapital = 1000
for jahr in range(1, 4):
    kapital = kapital * 1.07
    print(f"Jahr {jahr}: {kapital:.2f} €")
````

## Aufgabe

MotoTec möchte monatlich einen festen Betrag in die Maschinenrücklage einzahlen.
Der Zinssatz beträgt 7 % pro Jahr, der Anlagezeitraum 50 Jahre.

**(a)** Lesen Sie die monatliche Sparrate per `int(input())` ein. Die erlaubten Optionen sind: 50, 100, 150, 200, 250, 300 €.

**(b)** Validieren Sie die Eingabe. Falls die eingegebene Rate nicht einer der erlaubten Optionen entspricht, geben Sie eine Fehlermeldung aus.

**(c)** Berechnen Sie die Kapitalentwicklung über 50 Jahre bei 7 % Jahreszins. Verwenden Sie die Formel:
```
guthaben = (guthaben + 12 * monatsrate) * (100 + zinssatz) / 100
```

**(d)** Geben Sie das Endguthaben als formatierten f-String aus (auf 2 Nachkommastellen gerundet).

````python
# === Aufgabe (a): Monatliche Sparrate einlesen ===
guthaben = 0
zinssatz = 7
jahre = 50

print("Wie viel möchten Sie monatlich sparen? 50€, 100€, 150€, 200€, 250€, 300€")
monatsrate = int(input())

# === Aufgabe (b): Eingabe validieren ===
if (monatsrate == 50 or monatsrate == 100 or monatsrate == 150
        or monatsrate == 200 or monatsrate == 250 or monatsrate == 300):

    # === Aufgabe (c): Kapitalentwicklung über 50 Jahre berechnen ===
    for i in range(1, jahre + 1):
        guthaben = (guthaben + 12 * monatsrate) * (100 + zinssatz) / 100

    # === Aufgabe (d): Endguthaben ausgeben ===
    print(f"Das Guthaben der Maschinenrücklage nach {jahre} Jahren beträgt: {guthaben:.2f} €")
else:
    print("Falsche Eingabe! Bitte wählen Sie eine der erlaubten Raten: 50, 100, 150, 200, 250 oder 300 €.")
````
