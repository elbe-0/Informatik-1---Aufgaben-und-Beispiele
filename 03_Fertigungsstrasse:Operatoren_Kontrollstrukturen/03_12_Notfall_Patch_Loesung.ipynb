## Beispiele

### Häufige Syntaxfehler in Python

Wenn Sie Code aus anderen Sprachen nach Python übertragen, schleichen sich typische Fehler ein.
Hier einige Beispiele:

````python
# Beispiel 1: Doppelpunkt nach Schleifen und Bedingungen
# FALSCH: while (x > 0)
# RICHTIG:
x = 3
while x > 0:
    print(x)
    x -= 1
````

````python
# Beispiel 2: Vergleich (==) vs. Zuweisung (=)
# FALSCH: if (wert = 10):
# RICHTIG:
wert = 10
if wert == 10:
    print("Wert ist 10")
````

## Aufgabe

Der folgende Code enthält mehrere **Syntax-** und **Logikfehler**.
Die Funktion `pruefe_chargen` soll alle Chargennummern in einem Bereich prüfen und diejenigen ausgeben,
die eine besondere Qualitätsprüfung benötigen (Primzahl-Chargen).

**(a)** Finden Sie alle **Syntaxfehler** im Code. Notieren Sie die Zeile und beschreiben Sie den Fehler.

**(b)** Finden Sie alle **Logikfehler** im Code. Notieren Sie die Zeile und beschreiben Sie den Fehler.

**(c)** Korrigieren Sie alle Fehler und führen Sie den korrigierten Code aus.

### Fehleranalyse des Originalcodes

```python
def pruefe_chargen(start_charge, end_charge):
    l = 0  # Hilfsvariable l                          # FEHLER 5: l = 0 -> Endlosschleife (Logik)

    int n = start_charge                               # FEHLER 1: C-Style Typdeklaration (Syntax)
    while (n <= end_charge)                             # FEHLER 2: Fehlender Doppelpunkt (Syntax)
        zaehler = 0;                                    # FEHLER 3: Semikolon - C-Style (Syntax)
        i = 2
        while (i < n):
            if (n % i == 0):
                zaehler += 1
            i += 1
        if (zaehler = 0):                               # FEHLER 4: = statt == (Syntax/Logik)
            print(n, "ist eine Primzahl-Charge ...")
        n += l                                          # FEHLER 5: n += 0 -> Endlosschleife (Logik)
    return 0                                            # FEHLER 6: Unnötiger C-Style Return (Stil)
```

**Syntaxfehler:**
1. `int n = start_charge` - Typdeklaration wie in C/Java. Korrekt: `n = start_charge`
2. `while (n <= end_charge)` - Fehlender Doppelpunkt am Zeilenende. Korrekt: `while (n <= end_charge):`
3. `zaehler = 0;` - Semikolon am Zeilenende (in Python zwar gültig, aber C-Style und unüblich)
4. `if (zaehler = 0):` - Zuweisung statt Vergleich. Korrekt: `if (zaehler == 0):`

**Logikfehler:**
5. `l = 0` und `n += l` - Die Variable `l` ist 0, daher wird `n` nie erhöht, was eine Endlosschleife verursacht. Korrekt: `n += 1`
6. `return 0` - Unnötiger Rückgabewert im C-Stil. Kann entfernt werden.

````python
# === Aufgabe (c): Korrigierter Code ===

def pruefe_chargen(start_charge, end_charge):
    n = start_charge                          # FIX 1: Typdeklaration 'int' entfernt
    while (n <= end_charge):                   # FIX 2: Doppelpunkt ergänzt
        zaehler = 0                            # FIX 3: Semikolon entfernt
        i = 2
        while (i < n):
            if (n % i == 0):
                zaehler += 1
            i += 1
        if (zaehler == 0):                     # FIX 4: = durch == ersetzt
            print(n, "ist eine Primzahl-Charge und benötigt Qualitätsprüfung")
        n += 1                                 # FIX 5: l durch 1 ersetzt (Endlosschleife behoben)
                                               # FIX 6: 'return 0' entfernt

start = 11
ende = 50
pruefe_chargen(start, ende)
````
