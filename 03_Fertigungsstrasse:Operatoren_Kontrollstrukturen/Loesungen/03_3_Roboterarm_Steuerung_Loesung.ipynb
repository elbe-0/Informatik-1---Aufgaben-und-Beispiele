## Beispiele

Mit `if/elif/else` können mehrere Bedingungen nacheinander geprüft werden. Führen Sie die folgenden Zellen aus:

````python
# Einfache if/else-Verzweigung
temperatur = 85

if temperatur > 100:
    print("ALARM: Überhitzung!")
else:
    print("Temperatur im Normalbereich.")
````

Mit `elif` werden mehrere Bedingungen der Reihe nach geprüft. Sobald eine Bedingung `True` ist, wird der zugehörige Block ausgeführt und alle weiteren `elif`/`else`-Zweige werden übersprungen:

````python
# if/elif/else-Kette
drehzahl = 3500

if drehzahl > 5000:
    print("Drehzahl kritisch - Notabschaltung!")
elif drehzahl > 3000:
    print("Drehzahl erhöht - Überwachung aktiv.")
elif drehzahl > 1000:
    print("Drehzahl normal.")
else:
    print("Drehzahl zu niedrig - Maschine prüfen.")
````

Die Einrückung bestimmt, welche Anweisungen zu welchem Block gehören:

````python
# Einrückung entscheidet über die Zugehörigkeit
status = "aktiv"

if status == "aktiv":
    print("Maschine läuft.")
    print("Produktion gestartet.")  # gehört zum if-Block

print("Prüfung abgeschlossen.")  # wird IMMER ausgeführt
````

## Aufgabe

Der Roboterarm wird über `if/elif`-Ketten gesteuert. Analysieren Sie den Code, **bevor** Sie ihn ausführen.

### (a) Welches Fach wählt der Roboterarm?

Der folgende Code steuert den Roboterarm anhand des Teilegewichts. Das aktuelle Teil wiegt **40 kg**.

Lesen Sie den Code und beantworten Sie **vor dem Ausführen**:
- In welches Fach legt der Arm das Teil?
- Welche Bedingung greift zuerst?
- Warum wird die Bedingung `teil_gewicht % 10 == 0` (glatte Zehnerzahl) **nicht** erreicht?

````python
# === Aufgabe (a): if/elif - Die erste zutreffende Bedingung gewinnt ===
# teil_gewicht = 40
#
# Prüfung der Reihe nach:
# 1. teil_gewicht >= 40  ->  40 >= 40  ->  True  ->  DIESER Zweig wird ausgeführt!
# 2. teil_gewicht >= 20  ->  wird NICHT mehr geprüft (weil 1. schon True war)
# 3. teil_gewicht % 10 == 0  ->  wird NICHT mehr geprüft
# 4. else  ->  wird NICHT mehr geprüft
#
# Ergebnis: "Schwerlast-Ablage"
# Obwohl 40 % 10 == 0 ebenfalls True wäre, wird die Sonderablage
# nie erreicht, weil die Bedingung >= 40 weiter oben steht und
# bereits greift. Bei if/elif zählt nur der ERSTE Treffer.

teil_gewicht = 40

if teil_gewicht >= 40:
    fach = "Schwerlast-Ablage"
elif teil_gewicht >= 20:
    fach = "Standard-Ablage"
elif teil_gewicht % 10 == 0:
    fach = "Sonderablage (glatte Zehnerzahl)"
else:
    fach = "Leichtteile-Ablage"

print(f"Teil ({teil_gewicht} kg) -> {fach}")
````

**Erklärung zu (a):**

Bei einer `if/elif`-Kette wird **nur der erste zutreffende Zweig** ausgeführt. Alle nachfolgenden Zweige werden übersprungen - auch wenn deren Bedingung ebenfalls `True` wäre.

Hier ist `teil_gewicht >= 40` die erste Bedingung, die für `40` zutrifft. Deshalb landet das Teil in der Schwerlast-Ablage. Die Bedingung `teil_gewicht % 10 == 0` (40 ist eine glatte Zehnerzahl) wäre zwar auch wahr, wird aber gar nicht mehr geprüft.

Das ist der Grund, warum der Roboterarm "daneben greift": Die Reihenfolge der Bedingungen bestimmt das Verhalten. Wer die Sonderablage für glatte Zehnerzahlen nutzen will, müsste diese Bedingung **vor** den Gewichtsprüfungen platzieren.

### (b) Zwei Roboterarm-Programme - gleicher Code, andere Einrückung

Die folgenden zwei Code-Blöcke sehen fast identisch aus. Der einzige Unterschied ist die **Einrückung** einer einzigen Zeile.

Lesen Sie beide Varianten und notieren Sie **vor dem Ausführen**, was jeweils ausgegeben wird.

````python
# === Aufgabe (b) Variante 1 ===
# position = 100, greifer = "offen"
#
# if position > 80:          ->  100 > 80  ->  True
#     print("Arm fährt...")   ->  wird ausgeführt (im if-Block)
#     position = position + 50  ->  wird ausgeführt (im if-Block!)
#                                   position ist jetzt 150
# print(f"Position: ...")     ->  wird IMMER ausgeführt (außerhalb des if-Blocks)
#
# Ausgabe:
# Arm fährt zur Ablage.
# Position: 150, Greifer: offen

position = 100
greifer = "offen"

if position > 80:
    print("Arm fährt zur Ablage.")
    position = position + 50
print(f"Position: {position}, Greifer: {greifer}")
````

````python
# === Aufgabe (b) Variante 2 ===
# position = 100, greifer = "offen"
#
# if position > 80:          ->  100 > 80  ->  True
#     print("Arm fährt...")   ->  wird ausgeführt (im if-Block)
# position = position + 50    ->  wird IMMER ausgeführt (NICHT im if-Block!)
#                                 position ist jetzt 150
# print(f"Position: ...")     ->  wird IMMER ausgeführt
#
# Ausgabe:
# Arm fährt zur Ablage.
# Position: 150, Greifer: offen
#
# ACHTUNG: Bei position = 50 wäre der Unterschied sichtbar!
# Variante 1: position bleibt 50 (position + 50 ist im if-Block, der nicht ausgeführt wird)
# Variante 2: position wird 100 (position + 50 ist AUSSERHALB des if-Blocks, wird IMMER ausgeführt)

position = 100
greifer = "offen"

if position > 80:
    print("Arm fährt zur Ablage.")
position = position + 50
print(f"Position: {position}, Greifer: {greifer}")
````

**Erklärung zu (b):**

Beide Varianten erzeugen bei `position = 100` zufällig die **gleiche** Ausgabe, weil die Bedingung `position > 80` in beiden Fällen `True` ist.

Der entscheidende Unterschied zeigt sich, wenn die Bedingung `False` wäre (z.B. bei `position = 50`):

| | Variante 1 | Variante 2 |
|---|---|---|
| `position + 50` | Nur wenn `position > 80` | **Immer** |
| Bei `position = 50` | Position bleibt 50 | Position wird 100 |

In **Variante 1** ist `position = position + 50` eingerückt und gehört zum `if`-Block. Die Zeile wird nur ausgeführt, wenn die Bedingung `True` ist.

In **Variante 2** ist `position = position + 50` **nicht** eingerückt. Sie liegt außerhalb des `if`-Blocks und wird **immer** ausgeführt, unabhängig von der Bedingung.

Genau solche Einrückungsfehler können dazu führen, dass der Roboterarm bei bestimmten Teilen unerwartet die Position ändert.

````python
# === Beweis: Mit position = 50 wird der Unterschied sichtbar ===

print("--- Variante 1 (position + 50 eingerückt) ---")
position = 50
greifer = "offen"
if position > 80:
    print("Arm fährt zur Ablage.")
    position = position + 50
print(f"Position: {position}, Greifer: {greifer}")

print()

print("--- Variante 2 (position + 50 NICHT eingerückt) ---")
position = 50
greifer = "offen"
if position > 80:
    print("Arm fährt zur Ablage.")
position = position + 50
print(f"Position: {position}, Greifer: {greifer}")
````

### (c) Roboterarm reparieren

Der folgende Code soll den Roboterarm so steuern:
- Gewicht >= 50 kg: Schwerlast-Greifer verwenden, Ablage links
- Gewicht >= 20 kg: Standard-Greifer verwenden, Ablage mitte
- Sonst: Leicht-Greifer verwenden, Ablage rechts

**Nach** dem Greifen soll **immer** die Meldung `"Teil gegriffen."` ausgegeben werden.

Der Code enthält einen Einrückungsfehler. Finden und korrigieren Sie ihn.

````python
# === Aufgabe (c): Korrigierter Code ===
# Der Fehler: print("Teil gegriffen.") war in den else-Block eingerückt.
# Dadurch wurde die Meldung NUR bei Leichtteilen ausgegeben.
# Die Zeile muss AUSSERHALB der if/elif/else-Kette stehen,
# damit sie IMMER ausgeführt wird - egal welcher Greifer gewählt wurde.

teil_gewicht = 35

if teil_gewicht >= 50:
    greifer = "Schwerlast"
    ablage = "links"
elif teil_gewicht >= 20:
    greifer = "Standard"
    ablage = "mitte"
else:
    greifer = "Leicht"
    ablage = "rechts"
print("Teil gegriffen.")

print(f"Greifer: {greifer}, Ablage: {ablage}")
````

**Erklärung zu (c):**

Im fehlerhaften Code war `print("Teil gegriffen.")` im `else`-Block eingerückt:

```python
else:
    greifer = "Leicht"
    ablage = "rechts"
    print("Teil gegriffen.")  # <-- nur bei Leichtteilen!
```

Dadurch wurde `"Teil gegriffen."` nur ausgegeben, wenn das Teil in die Leichtteile-Ablage kam. Für Standard- und Schwerlast-Teile fehlte die Bestätigung.

Die Korrektur: `print("Teil gegriffen.")` wird auf die gleiche Ebene wie `if` gerückt - also **ohne Einrückung**. Damit liegt die Anweisung außerhalb der gesamten `if/elif/else`-Kette und wird bei jedem Durchlauf ausgeführt:

```python
else:
    greifer = "Leicht"
    ablage = "rechts"
print("Teil gegriffen.")  # <-- immer, unabhängig vom Zweig
```
