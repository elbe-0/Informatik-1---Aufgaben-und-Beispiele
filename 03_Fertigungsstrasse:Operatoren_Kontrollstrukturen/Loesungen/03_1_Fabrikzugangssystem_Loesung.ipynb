## Beispiele

Bevor Sie die Aufgabe lösen, machen Sie sich mit `while`-Schleifen und f-Strings vertraut. Führen Sie die folgenden Zellen aus:

Eine `while`-Schleife wiederholt einen Codeblock, solange eine Bedingung `True` ist:

```python
zaehler = 1
while zaehler <= 5:
    print(f"Durchlauf {zaehler}")
    zaehler = zaehler + 1

print("Schleife beendet.")
```

Mit `!=` (ungleich) können Sie eine Schleife laufen lassen, bis ein bestimmter Wert erreicht wird:

```python
farbe = ""
while farbe != "rot":
    farbe = input("Welche Farbe hat das Warnlicht? ")
    print(f"Sie haben '{farbe}' eingegeben.")

print("Richtig! Das Warnlicht ist rot.")
```

Mit f-Strings können Sie Variablen direkt in einen Text einbetten. Setzen Sie ein `f` vor den String und die Variable in geschweifte Klammern:

```python
maschine = "Presswerk"
temperatur = 87.5
print(f"Maschine: {maschine}, Temperatur: {temperatur} Grad")
```

Eine `while`-Schleife kann auch mit `and` mehrere Bedingungen gleichzeitig prüfen:

```python
eingabe = ""
runden = 0
while eingabe != "stop" and runden < 3:
    eingabe = input("Eingabe (oder 'stop'): ")
    runden = runden + 1
    print(f"Runde {runden} abgeschlossen.")

print("Fertig.")
```

## Aufgabe

Programmieren Sie eine Notfall-Zugangskontrolle für die MotoTec-Fabrik:

**(a)** Fragen Sie mit einer `while`-Schleife ein Passwort ab. Das korrekte Passwort lautet `"MotoTec2024"`. Solange das eingegebene Passwort falsch ist, soll die Schleife die Abfrage wiederholen.

**(b)** Begrenzen Sie die Anzahl der Fehlversuche auf maximal 5. Nach dem fünften falschen Versuch soll die Schleife abbrechen.

**(c)** Zeigen Sie nach jeder Fehleingabe die Anzahl der verbleibenden Versuche an. Verwenden Sie dafür einen f-String.

**(d)** Geben Sie nach der Schleife aus, ob der Zugang gewährt oder das System gesperrt wurde. Prüfen Sie dazu, ob das Passwort korrekt eingegeben wurde oder ob die maximale Anzahl an Versuchen erreicht ist.

```python
# === Aufgabe (a): Passwort-Abfrage mit while-Schleife ===
# Das korrekte Passwort wird als Variable gespeichert.
# Die Schleife läuft, solange das Passwort falsch ist.
korrektes_passwort = "MotoTec2024"
eingabe = ""

# === Aufgabe (b): Maximale Fehlversuche begrenzen ===
# Ein Zähler zählt die Versuche. Die Schleife prüft beide Bedingungen:
# Passwort falsch UND Versuche noch nicht aufgebraucht.
max_versuche = 5
versuche = 0

while eingabe != korrektes_passwort and versuche < max_versuche:
    eingabe = input("Passwort eingeben: ")
    versuche = versuche + 1

    # === Aufgabe (c): Verbleibende Versuche anzeigen ===
    # Nur bei Fehleingabe die verbleibenden Versuche ausgeben.
    if eingabe != korrektes_passwort:
        verbleibend = max_versuche - versuche
        print(f"Falsches Passwort! Verbleibende Versuche: {verbleibend}")

# === Aufgabe (d): Ergebnis nach der Schleife ===
# Nach der Schleife gibt es zwei mögliche Ausgänge:
# 1. Das Passwort wurde korrekt eingegeben -> Zugang gewährt
# 2. Alle Versuche aufgebraucht -> System gesperrt
if eingabe == korrektes_passwort:
    print("Zugang gewährt. Willkommen bei MotoTec.")
else:
    print("System gesperrt! Kontaktieren Sie die IT-Abteilung.")
```
