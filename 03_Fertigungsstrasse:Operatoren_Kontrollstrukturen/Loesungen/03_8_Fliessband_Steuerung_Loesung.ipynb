## Beispiele

Mit `break` wird eine Schleife sofort komplett beendet. Mit `continue` wird der aktuelle Schleifendurchlauf übersprungen und direkt mit dem nächsten Durchlauf weitergemacht.

```python
# Beispiel 1: break beendet die Schleife
for i in range(10):
    print(i)
    if i == 3:
        break
# Ausgabe: 0, 1, 2, 3 - danach wird die Schleife beendet
```

```python
# Beispiel 2: continue überspringt den Rest des Durchlaufs
for i in range(6):
    if i == 3:
        continue
    print(i)
# Ausgabe: 0, 1, 2, 4, 5 - die 3 wird übersprungen
```

### **(a)** Snippet 1: Qualitätsprüfung am Fließband

Das Fließband prüft Autos der Reihe nach. Bei einem defekten Auto wird das Band gestoppt.

**Vorhersage:** Was gibt dieser Code aus?

```python
autos = ["Auto-A1", "Auto-A2", "Auto-A3", "Auto-A4", "Auto-A5"]
defekt = "Auto-A3"

for auto in autos:
    if auto == defekt:
        print(f"STOPP! {auto} ist defekt - Band angehalten!")
        break
    print(f"{auto} geprüft - OK")

# === Aufgabe (a): Erwartete Ausgabe ===
# Auto-A1 geprüft - OK
# Auto-A2 geprüft - OK
# STOPP! Auto-A3 ist defekt - Band angehalten!
#
# Erklärung: Die Schleife läuft durch A1 und A2 normal durch.
# Bei A3 greift die if-Bedingung: Die Stopp-Meldung wird ausgegeben,
# dann beendet break die gesamte Schleife. A4 und A5 werden nie erreicht.
```

---

### **(b)** Snippet 2: Wartungsautos überspringen

Bestimmte Autos sind für die Wartung markiert und sollen auf dem Fließband übersprungen werden.

**Vorhersage:** Was gibt dieser Code aus?

```python
produktionsnummern = [101, 102, 103, 104, 105, 106]
wartung = [102, 105]

for nr in produktionsnummern:
    if nr in wartung:
        print(f"Nr. {nr} - Wartung, wird übersprungen")
        continue
    print(f"Nr. {nr} - Montage durchgeführt")

# === Aufgabe (b): Erwartete Ausgabe ===
# Nr. 101 - Montage durchgeführt
# Nr. 102 - Wartung, wird übersprungen
# Nr. 103 - Montage durchgeführt
# Nr. 104 - Montage durchgeführt
# Nr. 105 - Wartung, wird übersprungen
# Nr. 106 - Montage durchgeführt
#
# Erklärung: Die Schleife läuft durch alle 6 Nummern.
# Bei Nr. 102 und 105 greift continue: Die Wartungs-Meldung wird ausgegeben,
# dann springt die Schleife direkt zum nächsten Durchlauf.
# Die Montage-Zeile wird für diese Nummern übersprungen.
# Wichtig: Die Schleife läuft weiter - sie wird NICHT beendet.
```

### **(c)** Snippet 3: Produktionszähler mit Notabschaltung

Das Fließband produziert in einer `while`-Schleife. Bei bestimmten Qualitätswerten wird das Auto übersprungen, bei einem kritischen Fehler stoppt das Band.

**Vorhersage:** Was gibt dieser Code aus?

```python
position = 0
qualitaet = [85, 92, 40, 78, 15, 95, 88]

while position < len(qualitaet):
    wert = qualitaet[position]
    position += 1

    if wert < 30:
        print(f"Position {position}: Qualität {wert} - KRITISCH! Band gestoppt!")
        break

    if wert < 50:
        print(f"Position {position}: Qualität {wert} - mangelhaft, übersprungen")
        continue

    print(f"Position {position}: Qualität {wert} - freigegeben")

# === Aufgabe (c): Erwartete Ausgabe ===
# Position 1: Qualität 85 - freigegeben
# Position 2: Qualität 92 - freigegeben
# Position 3: Qualität 40 - mangelhaft, übersprungen
# Position 4: Qualität 78 - freigegeben
# Position 5: Qualität 15 - KRITISCH! Band gestoppt!
#
# Erklärung: position wird VOR der Prüfung erhöht (daher startet die Ausgabe bei 1).
# - Position 1 (85) und 2 (92): Qualität >= 50, also freigegeben.
# - Position 3 (40): Qualität < 50 aber >= 30, also mangelhaft -> continue überspringt die Freigabe.
# - Position 4 (78): Qualität >= 50, also freigegeben.
# - Position 5 (15): Qualität < 30, also KRITISCH -> break stoppt die gesamte Schleife.
# - Position 6 (95) und 7 (88) werden nie erreicht.
```

### **(d)** Reflexion

**Antworten:**

- **Was ist der Unterschied zwischen `break` und `continue`?**  
  `break` beendet die gesamte Schleife sofort - kein weiterer Durchlauf wird ausgeführt. `continue` überspringt nur den Rest des *aktuellen* Durchlaufs und macht mit dem nächsten Schleifendurchlauf weiter.

- **In Snippet 3: Warum wird Position 6 (Qualität 95) nie erreicht?**  
  Weil bei Position 5 die Qualität 15 beträgt, was unter dem kritischen Schwellenwert von 30 liegt. Dadurch wird `break` ausgeführt, was die gesamte `while`-Schleife sofort beendet. Alle nachfolgenden Positionen (6 und 7) werden nicht mehr verarbeitet.
