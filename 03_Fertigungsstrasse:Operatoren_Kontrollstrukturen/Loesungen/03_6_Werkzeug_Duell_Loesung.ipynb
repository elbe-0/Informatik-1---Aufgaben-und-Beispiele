## Beispiele

Mit `random.randint()` können Sie zufällige Ganzzahlen erzeugen. Mit verschachtelten `if/elif`-Bedingungen können Sie verschiedene Kombinationen auswerten.

```python
# Beispiel 1: Zufällige Zahl erzeugen
import random

zufallszahl = random.randint(1, 3)
print(f"Zufällige Zahl zwischen 1 und 3: {zufallszahl}")
```

```python
# Beispiel 2: Verschachtelte Bedingungen für zwei Wahlen
wahl_a = 2
wahl_b = 1

if wahl_a == 1:
    if wahl_b == 1:
        print("Beide gleich!")
    elif wahl_b == 2:
        print("A verliert gegen B")
elif wahl_a == 2:
    if wahl_b == 1:
        print("A gewinnt gegen B")
    elif wahl_b == 2:
        print("Beide gleich!")
```

## Aufgabe

Programmieren Sie das Werkzeug-Duell mit fünf Werkzeugen und asymmetrischen Regeln.

**Die fünf Werkzeuge:**
1. Schraubenschlüssel
2. Schaltplan
3. Sicherung
4. Kabel
5. Multimeter

**Die Gewinnregeln (jedes Werkzeug schlägt genau zwei andere):**

| Werkzeug | schlägt | Begründung |
|---|---|---|
| Schraubenschlüssel | Sicherung, Multimeter | löst Sicherung heraus, zerstört Multimeter |
| Schaltplan | Schraubenschlüssel, Kabel | leitet Schraubenschlüssel-Einsatz, dokumentiert Kabelführung |
| Sicherung | Schaltplan, Kabel | schützt Schaltplan vor Überspannung, unterbricht Kabel |
| Kabel | Schraubenschlüssel, Multimeter | blockiert Schraubenschlüssel, überbrückt Multimeter |
| Multimeter | Schaltplan, Sicherung | prüft Schaltplan auf Fehler, testet Sicherung |

**(a)** Zeigen Sie das Menü der 5 Werkzeuge an und lesen Sie die Wahl des Spielers ein.  
**(b)** Erzeugen Sie eine zufällige Computer-Wahl (1-5).  
**(c)** Geben Sie aus, welches Werkzeug der Computer gewählt hat.  
**(d)** Werten Sie alle möglichen Kombinationen aus und geben Sie das Ergebnis aus: `"Gewonnen!"`, `"Verloren!"` oder `"Unentschieden!"`

```python
import random

# === Aufgabe (a): Menü anzeigen und Spieler-Wahl einlesen ===
print("Bitte wählen Sie ein Werkzeug:")
print("1 - Schraubenschlüssel")
print("2 - Schaltplan")
print("3 - Sicherung")
print("4 - Kabel")
print("5 - Multimeter")
spieler_wahl = int(input())

# === Aufgabe (b): Zufällige Computer-Wahl erzeugen ===
computer_wahl = random.randint(1, 5)

# === Aufgabe (c): Computer-Wahl ausgeben ===
print(f"Der Computer hat {computer_wahl} gewählt.")

# === Aufgabe (d) und (e): Verschachtelte if/elif-Struktur ===
if computer_wahl == 1:  # Computer: Schraubenschlüssel
    if spieler_wahl == 1:
        print("Unentschieden!")
    elif spieler_wahl == 2:
        print("Schaltplan leitet Schraubenschlüssel-Einsatz - Gewonnen!")
    elif spieler_wahl == 3:
        print("Schraubenschlüssel löst Sicherung heraus - Verloren!")
    elif spieler_wahl == 4:
        print("Kabel blockiert Schraubenschlüssel - Gewonnen!")
    elif spieler_wahl == 5:
        print("Schraubenschlüssel zerstört Multimeter - Verloren!")

elif computer_wahl == 2:  # Computer: Schaltplan
    if spieler_wahl == 1:
        print("Schaltplan leitet Schraubenschlüssel-Einsatz - Verloren!")
    elif spieler_wahl == 2:
        print("Unentschieden!")
    elif spieler_wahl == 3:
        print("Sicherung schützt Schaltplan vor Überspannung - Gewonnen!")
    elif spieler_wahl == 4:
        print("Schaltplan dokumentiert Kabelführung - Verloren!")
    elif spieler_wahl == 5:
        print("Multimeter prüft Schaltplan auf Fehler - Gewonnen!")

elif computer_wahl == 3:  # Computer: Sicherung
    if spieler_wahl == 1:
        print("Schraubenschlüssel löst Sicherung heraus - Gewonnen!")
    elif spieler_wahl == 2:
        print("Sicherung schützt Schaltplan vor Überspannung - Verloren!")
    elif spieler_wahl == 3:
        print("Unentschieden!")
    elif spieler_wahl == 4:
        print("Sicherung unterbricht Kabel - Verloren!")
    elif spieler_wahl == 5:
        print("Multimeter testet Sicherung - Gewonnen!")

elif computer_wahl == 4:  # Computer: Kabel
    if spieler_wahl == 1:
        print("Kabel blockiert Schraubenschlüssel - Verloren!")
    elif spieler_wahl == 2:
        print("Schaltplan dokumentiert Kabelführung - Gewonnen!")
    elif spieler_wahl == 3:
        print("Sicherung unterbricht Kabel - Gewonnen!")
    elif spieler_wahl == 4:
        print("Unentschieden!")
    elif spieler_wahl == 5:
        print("Kabel überbrückt Multimeter - Verloren!")

elif computer_wahl == 5:  # Computer: Multimeter
    if spieler_wahl == 1:
        print("Schraubenschlüssel zerstört Multimeter - Gewonnen!")
    elif spieler_wahl == 2:
        print("Multimeter prüft Schaltplan auf Fehler - Verloren!")
    elif spieler_wahl == 3:
        print("Multimeter testet Sicherung - Verloren!")
    elif spieler_wahl == 4:
        print("Kabel überbrückt Multimeter - Gewonnen!")
    elif spieler_wahl == 5:
        print("Unentschieden!")
```
