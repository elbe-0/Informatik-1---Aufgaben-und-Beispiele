## Beispiele

Für diese Aufgabe benötigen Sie zwei Operatoren: Ganzzahldivision (`//`) und Modulo (`%`). Führen Sie die folgenden Zellen aus:

Der Operator `//` teilt und schneidet die Nachkommastellen ab (Ganzzahldivision):

````python
print(17 // 5)
print(100 // 3)
print(42 // 10)
````

Der Operator `%` (Modulo) liefert den Rest einer Division:

````python
print(17 % 5)
print(100 % 3)
print(42 % 10)
````

Mit `int(input())` können Sie eine Zahl direkt als Ganzzahl einlesen:

````python
zahl = int(input("Geben Sie eine Zahl ein: "))
print("Das Doppelte ist:", zahl * 2)
````

## Aufgabe

Schreiben Sie ein Python-Programm zur Prüfung von dreistelligen Auto-Seriennummern:

**(a)** Lesen Sie eine dreistellige Seriennummer per `int(input())` ein und berechnen Sie die Quersumme (Summe der einzelnen Ziffern). Geben Sie die Quersumme als Prüfziffer aus, z.B.:

```
Seriennummer: 385
Prüfziffer (Quersumme): 16
```

**(b)** Prüfen Sie, ob die Seriennummer durch 5 teilbar ist. Falls ja, geben Sie `"Sondermodell"` aus, andernfalls `"Standardmodell"`.

**(c)** Prüfen Sie, ob die Seriennummer durch 5 teilbar **und** größer als 100 ist. Falls beide Bedingungen erfüllt sind, geben Sie `"Exportmodell"` aus, andernfalls `"Kein Exportmodell"`.

````python
# === Aufgabe (a): Seriennummer einlesen und Quersumme berechnen ===
# Die Seriennummer wird als Ganzzahl eingelesen.
seriennummer = int(input("Seriennummer (3-stellig): "))

# Mit // und % extrahieren wir die einzelnen Ziffern:
# Hunderter: 385 // 100 = 3
hunderter = seriennummer // 100

# Rest nach Abzug der Hunderter: 385 % 100 = 85
rest = seriennummer % 100

# Zehner: 85 // 10 = 8
zehner = rest // 10

# Einer: 85 % 10 = 5
einer = rest % 10

# Quersumme = Summe aller Ziffern
quersumme = hunderter + zehner + einer

print("Seriennummer:", seriennummer)
print("Prüfziffer (Quersumme):", quersumme)

# === Aufgabe (b): Teilbar durch 5? Sondermodell oder Standardmodell ===
# Eine Zahl ist durch 5 teilbar, wenn der Rest der Division 0 ergibt.
if seriennummer % 5 == 0:
    print("Sondermodell")
else:
    print("Standardmodell")

# === Aufgabe (c): Teilbar durch 5 UND größer als 100? Exportmodell ===
# Mit 'and' verknüpfen wir zwei Bedingungen.
# Beide müssen gleichzeitig wahr sein.
if seriennummer % 5 == 0 and seriennummer > 100:
    print("Exportmodell")
else:
    print("Kein Exportmodell")
````
