## Briefing

In der MotoTec-Fabrik arbeiten die Mitarbeiter in Schichten. Damit die Produktion reibungslos
läuft, müssen die Mitarbeiter in gleichgroße Teams aufgeteilt werden. Nicht jede beliebige
Teamgröße ist möglich – nur wenn die Anzahl der Mitarbeiter ohne Rest durch die Teamgröße
teilbar ist, funktioniert die Aufteilung. Ihre Aufgabe ist es, alle möglichen Teamaufteilungen
zu berechnen.

## Stakeholder-Dialog

*Tarek Özdal, der Schichtleiter, kommt mit einem Klemmbrett in der Hand auf Sie zu.*

> **Özdal:** „Wir haben eine bestimmte Anzahl Mitarbeiter pro Schicht. Ich muss wissen,
> in wie viele gleichgroße Teams ich die aufteilen kann. Nicht ungefähr – exakt. Jedes Team
> muss gleich groß sein, sonst gibt es Beschwerden. Schreiben Sie mir ein Programm, das
> alle möglichen Aufteilungen berechnet.“

## Beispiele

Mit einer `for`-Schleife und `range()` können Sie einen Codeblock wiederholt ausführen.
Der Modulo-Operator `%` liefert den Rest einer ganzzahligen Division.

```python
# Beispiel 1: for-Schleife mit range()
# range(start, stop, step) erzeugt Zahlen von start bis stop-1
for i in range(1, 6, 1):
    print("Durchlauf Nr.", i)
```

```python
# Beispiel 2: Modulo-Operator %
# Der Rest einer Division zeigt, ob eine Zahl teilbar ist
print(10 % 3)   # Ergebnis: 1 (10 geteilt durch 3 = 3 Rest 1)
print(10 % 5)   # Ergebnis: 0 (10 geteilt durch 5 = 2 Rest 0)
print(10 % 2)   # Ergebnis: 0 (10 ist durch 2 teilbar)
```

## Aufgabe

**(a)** Lesen Sie die Anzahl der Mitarbeiter per `int(input())` ein.

**(b)** Berechnen Sie mit einer `for`-Schleife und `range()` alle ganzzahligen Teiler
der Mitarbeiteranzahl (von 1 bis einschließlich der Anzahl selbst).

**(c)** Geben Sie für jeden Teiler aus: `"X Mitarbeiter = Y Teams à Z Personen"`

```python
# === Aufgabe (a): Anzahl der Mitarbeiter einlesen ===
print("Wie viele Mitarbeiter sind in der Schicht?")
mitarbeiter = int(input())

# === Aufgabe (b): Alle ganzzahligen Teiler berechnen ===
# === Aufgabe (c): Ausgabe der Teamaufteilung ===
for i in range(1, mitarbeiter + 1, 1):
    rest = mitarbeiter % i
    if rest == 0:
        teamgroesse = mitarbeiter // i
        print(mitarbeiter, "Mitarbeiter =", i, "Teams à", teamgroesse, "Personen")
```
