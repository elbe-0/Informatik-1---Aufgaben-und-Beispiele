## Briefing

In der MotoTec-Fertigung werden Stahlblöcke schrittweise halbiert, um dünnere Bleche
herzustellen. Dabei wird das Gewicht des Blocks in jedem Schritt ganzzahlig halbiert.
Irgendwann ist der Block so leicht, dass eine weitere Halbierung nicht mehr möglich ist,
weil das Gewicht ungerade wird. Ihre Aufgabe ist es zu berechnen, wie oft ein Stahlblock
halbiert werden kann, bevor er nicht mehr durch 2 teilbar ist.

## Stakeholder-Dialog

*Tarek Özdal steht an der Walzstraße und betrachtet einen massiven Stahlblock.*

> *„Wie dünn können wir das Blech schneiden? Wir halbieren den Block immer
> wieder, bis es nicht mehr geht. Ich brauche ein Programm, das mir zeigt, wie oft
> wir halbieren können und was nach jedem Schritt übrig bleibt. Rechnen Sie das aus.“*


## Beispiele

Eine `while`-Schleife wiederholt einen Codeblock, solange eine Bedingung erfüllt ist.
Mit `//` führen Sie eine ganzzahlige Division durch (das Ergebnis wird abgerundet).

```python
# Beispiel 1: while-Schleife mit Zähler
zaehler = 0
while zaehler < 5:
    print("Zählerstand:", zaehler)
    zaehler += 1
print("Fertig! Schleife lief", zaehler, "Mal.")
```

```python
# Beispiel 2: Ganzzahlige Division mit //
print(17 // 2)   # Ergebnis: 8 (abgerundet)
print(10 // 2)   # Ergebnis: 5
print(10 % 2)    # Ergebnis: 0 (kein Rest, also durch 2 teilbar)
print(17 % 2)    # Ergebnis: 1 (Rest 1, also nicht durch 2 teilbar)
```
