## Beispiele

Mit `input()` können Sie Texteingaben vom Benutzer einlesen. Da `input()` immer einen String zurückgibt, müssen Zahlen mit `float()` umgewandelt werden:

```python
# input() gibt immer einen String zurück
eingabe = input("Menge in kg: ")
print("Typ:", type(eingabe))
print("Wert:", eingabe)
```

```python
# Mit float() wird der String in eine Kommazahl umgewandelt
menge = float(input("Menge in kg: "))
print("Typ:", type(menge))
print("Das Doppelte:", menge * 2)
```

Mit `if/elif` können Sie Strings vergleichen und je nach Eingabe unterschiedliche Aktionen ausführen:

```python
# String-Vergleich mit if/elif
material = input("Material (Stahl/Aluminium): ")

if material == "Stahl":
    print("Dichte: 7.85 kg/l")
elif material == "Aluminium":
    print("Dichte: 2.70 kg/l")
else:
    print("Unbekanntes Material.")
```
