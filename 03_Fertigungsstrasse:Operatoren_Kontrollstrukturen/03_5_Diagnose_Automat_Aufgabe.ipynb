## Beispiele

Mit `random.randint()` können Sie eine zufällige Ganzzahl in einem Bereich erzeugen. Die Bibliothek `random` wurde bereits in Kapitel 2 eingeführt - hier eine kurze Auffrischung:

```python
import random

# Zufällige Ganzzahl zwischen 1 und 6 (wie ein Würfel)
wurf = random.randint(1, 6)
print("Würfelergebnis:", wurf)
```

```python
# Zufällige Ganzzahl zwischen 0 und 2
zahl = random.randint(0, 2)
print("Zufallszahl:", zahl)

# Je nach Zahl eine andere Aktion
if zahl == 0:
    print("Aktion A")
elif zahl == 1:
    print("Aktion B")
elif zahl == 2:
    print("Aktion C")
```
