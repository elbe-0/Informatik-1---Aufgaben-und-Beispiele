## Beispiele

Mit `break` wird eine Schleife sofort komplett beendet. Mit `continue` wird der aktuelle Schleifendurchlauf übersprungen und direkt mit dem nächsten Durchlauf weitergemacht.

```python
# Beispiel 1: break beendet die Schleife
for i in range(10):
    print(i)
    if i == 3:
        break
# Ausgabe: 0, 1, 2, 3 - danach wird die Schleife beendet
```

```python
# Beispiel 2: continue überspringt den Rest des Durchlaufs
for i in range(6):
    if i == 3:
        continue
    print(i)
# Ausgabe: 0, 1, 2, 4, 5 - die 3 wird übersprungen
```

### **(a)** Snippet 1: Qualitätsprüfung am Fließband

Das Fließband prüft Autos der Reihe nach. Bei einem defekten Auto wird das Band gestoppt.

**Vorhersage:** Was gibt dieser Code aus?

```python
autos = ["Auto-A1", "Auto-A2", "Auto-A3", "Auto-A4", "Auto-A5"]
defekt = "Auto-A3"

for auto in autos:
    if auto == defekt:
        print(f"STOPP! {auto} ist defekt - Band angehalten!")
        break
    print(f"{auto} geprüft - OK")
```

---

### **(b)** Snippet 2: Wartungsautos überspringen

Bestimmte Autos sind für die Wartung markiert und sollen auf dem Fließband übersprungen werden.

**Vorhersage:** Was gibt dieser Code aus?

```python
produktionsnummern = [101, 102, 103, 104, 105, 106]
wartung = [102, 105]

for nr in produktionsnummern:
    if nr in wartung:
        print(f"Nr. {nr} - Wartung, wird übersprungen")
        continue
    print(f"Nr. {nr} - Montage durchgeführt")
```

---

### **(c)** Snippet 3: Produktionszähler mit Notabschaltung

Das Fließband produziert in einer `while`-Schleife. Bei bestimmten Qualitätswerten wird das Auto übersprungen, bei einem kritischen Fehler stoppt das Band.

**Vorhersage:** Was gibt dieser Code aus?

```python
position = 0
qualitaet = [85, 92, 40, 78, 15, 95, 88]

while position < len(qualitaet):
    wert = qualitaet[position]
    position += 1

    if wert < 30:
        print(f"Position {position}: Qualität {wert} - KRITISCH! Band gestoppt!")
        break

    if wert < 50:
        print(f"Position {position}: Qualität {wert} - mangelhaft, übersprungen")
        continue

    print(f"Position {position}: Qualität {wert} - freigegeben")
```
