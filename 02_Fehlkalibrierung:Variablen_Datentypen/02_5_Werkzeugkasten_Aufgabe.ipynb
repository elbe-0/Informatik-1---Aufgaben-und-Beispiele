## `math` - Mathematische Funktionen

Mit `import math` stehen mathematische Standardfunktionen zur Verfügung:

```python
import math

# Toleranzberechnung: Wurzel aus der Varianz ergibt die Standardabweichung
varianz = 16.0
toleranz = math.sqrt(varianz)
print("Varianz:", varianz)
print("Toleranzgrenze (Standardabweichung):", toleranz)

# Kreisumfang einer Welle berechnen
radius = 5.0  # cm
umfang = 2 * math.pi * radius
print("Wellenumfang:", umfang, "cm")

# Aufrunden mit ceil()
materialverbrauch = 7.2  # Einheiten
bestellung = math.ceil(materialverbrauch)
print("Materialverbrauch:", materialverbrauch, "-> Bestellung:", bestellung, "Einheiten")
```

## `random` - Zufallszahlen

Für Testsimulationen und Stichproben brauchen wir Zufallszahlen:

```python
import random

# Zufällige Testtemperatur zwischen 50 und 90 Grad
test_temperatur = random.randint(50, 90)
print("Simulierte Temperatur:", test_temperatur, "°C")

# Nochmal ausführen - jedes Mal ein anderer Wert!
test_temperatur = random.randint(50, 90)
print("Simulierte Temperatur:", test_temperatur, "°C")
```

## `time` - Zeit und Verzögerungen

Für Logeinträge und Wartepausen:

```python
import time

# Aktuelle Zeit als Zeitstempel
zeitstempel = time.strftime("%d.%m.%Y %H:%M:%S", time.localtime())
print("Logeintrag:", zeitstempel, "- System gestartet")
```

```python
# Verzögerung im Programmablauf
print("Sensor wird kalibriert...")
time.sleep(2)
print("Kalibrierung abgeschlossen.")
```

```python
# Alles zusammen: Logeintrag mit Zeitstempel
zeitstempel = time.strftime("%H:%M:%S", time.localtime())
print("[" + zeitstempel + "] Temperatur:", random.randint(70, 85), "°C")

time.sleep(1)

zeitstempel = time.strftime("%H:%M:%S", time.localtime())
print("[" + zeitstempel + "] Temperatur:", random.randint(70, 85), "°C")
```
