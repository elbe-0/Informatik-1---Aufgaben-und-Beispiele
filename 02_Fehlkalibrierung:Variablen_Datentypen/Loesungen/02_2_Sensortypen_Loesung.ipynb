## Beispiele

In Python hat jeder Wert einen **Datentyp**. Mit `type()` können Sie den Typ einer Variable abfragen:

```python
wert = 25
print("Wert:", wert)
print("Typ:", type(wert))
```

```python
wert = 3.141
print("Wert:", wert)
print("Typ:", type(wert))
```

```python
wert = "Hallo"
print("Wert:", wert)
print("Typ:", type(wert))
```

## Aufgabe

Der Drehzahlsensor liefert den Wert `1500` (Umdrehungen pro Minute). Ein Korrekturfaktor von `2.0` soll die Drehzahl anpassen.

**(a)** Legen Sie eine Variable `drehzahl` mit dem Wert `1500` an. Geben Sie den Wert und den Datentyp mit `type()` aus.

**(b)** Dividieren Sie `drehzahl` durch den Korrekturfaktor `2.0` und speichern Sie das Ergebnis wieder in `drehzahl`. Geben Sie erneut Wert und Datentyp aus.

Was hat sich geändert? Warum?

**(c)** Testen Sie: Was passiert, wenn Sie stattdessen durch `2` (ohne Dezimalpunkt) dividieren? Ändert sich der Datentyp trotzdem?

```python
# === Aufgabe (a): Variable anlegen und Typ prüfen ===
# Der Sensor liefert eine Ganzzahl (int).
drehzahl = 1500
print("Drehzahl:", drehzahl)
print("Typ:", type(drehzahl))  # <class 'int'>
```

```python
# === Aufgabe (b): Division durch Korrekturfaktor 2.0 ===
# Wenn ein int durch einen float dividiert wird,
# wird das Ergebnis automatisch zu float.
drehzahl = 1500
drehzahl = drehzahl / 2.0
print("Drehzahl:", drehzahl)   # 750.0
print("Typ:", type(drehzahl))  # <class 'float'>

# Erkenntnis: Der Typ hat sich von int zu float geändert!
# Python "promoted" automatisch zu float, wenn mindestens
# ein Operand ein float ist.
```

```python
# === Aufgabe (c): Division durch 2 (Ganzzahl) ===
# Auch bei Division durch eine Ganzzahl wird das
# Ergebnis in Python immer float!
drehzahl = 1500
drehzahl = drehzahl / 2
print("Drehzahl:", drehzahl)   # 750.0
print("Typ:", type(drehzahl))  # <class 'float'>

# Erkenntnis: Der /-Operator liefert in Python IMMER
# einen float - egal ob die Operanden int oder float sind.
# Für Ganzzahl-Division gibt es den //-Operator:
drehzahl = 1500
drehzahl = drehzahl // 2
print("Drehzahl (Ganzzahl-Division):", drehzahl)  # 750
print("Typ:", type(drehzahl))  # <class 'int'>
```
