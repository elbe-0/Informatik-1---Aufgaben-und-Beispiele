Das Monitoring-System kann mit Text nicht rechnen - Grenzwertüberschreitungen werden nicht erkannt. Dr. Voss braucht eine Lösung, bevor die Maschinen überhitzen.

## Beispiele

Mit `float()`, `int()` und `str()` können Sie Werte gezielt in einen anderen Datentyp umwandeln:

````python
# String in float umwandeln
text_wert = "33.3"
print("Vorher:", type(text_wert))

zahl_wert = float(text_wert)
print("Nachher:", type(zahl_wert))
print("Wert:", zahl_wert)
````

## Aufgabe

Der Temperatursensor liefert den Wert `"72.4"` als String. Der Grenzwert für einen Alarm liegt bei `85.0` Grad.

**(a)** Legen Sie die Variable `sensor_wert = "72.4"` an. Prüfen Sie den Datentyp mit `type()` und geben Sie ihn aus.

**(b)** Wandeln Sie `sensor_wert` in eine Gleitkommazahl um und speichern Sie das Ergebnis in einer neuen Variable `temperatur`. Prüfen Sie auch hier den Datentyp.

**(c)** Berechnen Sie den Abstand zum Grenzwert und geben Sie eine Meldung aus, z.B.:

```
Aktuelle Temperatur: 72.4 °C
Abstand zum Grenzwert: 12.6 °C
```

**(d)** Wandeln Sie die berechnete Temperatur mit `int()` in eine Ganzzahl um. Was passiert mit den Nachkommastellen?

````python
# === Aufgabe (a): Sensor-Wert als String ===
# Der Sensor liefert den Wert als Text (String).
sensor_wert = "72.4"
print("Sensor-Wert:", sensor_wert)
print("Typ:", type(sensor_wert))  # <class 'str'>
````

````python
# === Aufgabe (b): In Gleitkommazahl umwandeln ===
# float() wandelt den String "72.4" in die Zahl 72.4 um.
temperatur = float(sensor_wert)
print("Temperatur:", temperatur)
print("Typ:", type(temperatur))  # <class 'float'>
````

````python
# === Aufgabe (c): Abstand zum Grenzwert berechnen ===
# Jetzt können wir mit dem Wert rechnen.
grenzwert = 85.0
abstand = grenzwert - temperatur

print("Aktuelle Temperatur:", temperatur, "°C")
print("Abstand zum Grenzwert:", abstand, "°C")
````

````python
# === Aufgabe (d): In Ganzzahl umwandeln ===
# int() schneidet die Nachkommastellen ab (kein Runden!).
temperatur_ganzzahl = int(temperatur)
print("Als Ganzzahl:", temperatur_ganzzahl)  # 72 (nicht 72.4!)
print("Typ:", type(temperatur_ganzzahl))     # <class 'int'>

# Achtung: int() rundet NICHT, sondern schneidet ab!
# int(72.9) ergibt 72, nicht 73.
````
