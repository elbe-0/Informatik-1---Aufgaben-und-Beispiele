## Beispiele

Variablen speichern Werte. Der Datentyp wird in Python automatisch durch den zugewiesenen Wert festgelegt:

````python
# Ganzzahl (int)
anzahl = 33

# Gleitkommazahl (float)
summe = 3.5
ergebnis = 10.0   # auch mit .0 ist es ein float

# Text (str)
frage = "Wie geht es?"
antwort = 'gut'

print(anzahl, type(anzahl))
print(summe, type(summe))
print(frage, type(frage))
````

## Aufgabe

Rekonstruieren Sie den Maschinenpass für das **Presswerk Alpha** anhand der Papierunterlagen:

| Eigenschaft | Wert | Datentyp |
|---|---|---|
| Maschinen-ID | 1042 | Ganzzahl |
| Maximaltemperatur | 85.5 | Gleitkommazahl |
| Bezeichnung | Presswerk Alpha | Text |
| Baujahr | 2019 | Ganzzahl |
| Leistung (kW) | 220.0 | Gleitkommazahl |

**(a)** Legen Sie für jede Eigenschaft eine Variable mit einem passenden Namen und dem korrekten Wert an.

**(b)** Geben Sie alle Variablen mit `print()` aus, z.B.:

```
=== Maschinenpass ===
ID: 1042
Bezeichnung: Presswerk Alpha
Maximaltemperatur: 85.5 °C
Baujahr: 2019
Leistung: 220.0 kW
```

**(c)** Überprüfen Sie mit `type()`, ob jede Variable den erwarteten Datentyp hat.

````python
# === Aufgabe (a): Variablen anlegen ===
# Jede Variable bekommt den passenden Datentyp:
# Ganzzahlen (int) für ID und Baujahr,
# Gleitkommazahlen (float) für Temperatur und Leistung,
# Text (str) für die Bezeichnung.

maschinen_id = 1042
max_temp = 85.5
bezeichnung = "Presswerk Alpha"
baujahr = 2019
leistung_kw = 220.0
````

````python
# === Aufgabe (b): Maschinenpass ausgeben ===
print("=== Maschinenpass ===")
print("ID:", maschinen_id)
print("Bezeichnung:", bezeichnung)
print("Maximaltemperatur:", max_temp, "°C")
print("Baujahr:", baujahr)
print("Leistung:", leistung_kw, "kW")
````

````python
# === Aufgabe (c): Datentypen prüfen ===
print("maschinen_id:", type(maschinen_id))   # <class 'int'>
print("max_temp:", type(max_temp))             # <class 'float'>
print("bezeichnung:", type(bezeichnung))       # <class 'str'>
print("baujahr:", type(baujahr))               # <class 'int'>
print("leistung_kw:", type(leistung_kw))       # <class 'float'>
````
