## Beispiele

Mit dem Zeichen `#` können Sie Teile des Codes von der Ausführung ausschließen. Alles nach `#` in einer Zeile wird von Python ignoriert:

```python
# Diese Zeile wird nicht ausgeführt
print("Diese Zeile wird ausgeführt")
# print("Diese Zeile wird NICHT ausgeführt")
print("Diese Zeile auch")  # Kommentar am Zeilenende
```

Kommentare können auch dazu dienen, Code zu **dokumentieren** - also zu erklären, was der Code tut:

```python
# Programm: Temperaturumrechnung
# Autor: IT-Team MotoTec
# Datum: 2024-10-15

# Temperatur in Celsius
temperatur_c = 85.5

# Umrechnung in Fahrenheit
temperatur_f = temperatur_c * 9/5 + 32

print("Temperatur:", temperatur_f, "°F")
```
