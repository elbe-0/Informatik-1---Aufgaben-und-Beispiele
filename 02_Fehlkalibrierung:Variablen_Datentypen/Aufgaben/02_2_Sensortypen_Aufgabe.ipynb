## Beispiele

In Python hat jeder Wert einen **Datentyp**. Mit `type()` können Sie den Typ einer Variable abfragen:

```python
wert = 25
print("Wert:", wert)
print("Typ:", type(wert))
```

```python
wert = 3.141
print("Wert:", wert)
print("Typ:", type(wert))
```

```python
wert = "Hallo"
print("Wert:", wert)
print("Typ:", type(wert))
```
