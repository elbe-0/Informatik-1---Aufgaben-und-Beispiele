## Aufgabe

Simulieren Sie einen Stresstest: Messen Sie 10 Temperaturwerte und stellen Sie diese in einem Diagramm dar.

**(a)** Erzeugen Sie in einer `while`-Schleife **10 zufällige Temperaturwerte** zwischen 50 und 90 Grad (`random.randint(50, 90)`).

**(b)** Ermitteln Sie für jeden Messwert die aktuelle Uhrzeit mit `time.strftime("%H:%M:%S", time.localtime())`.

**(c)** Geben Sie jeden Messwert mit Uhrzeit auf dem Bildschirm aus.

**(d)** Tragen Sie jeden Messwert mit `plt.plot()` in ein Diagramm ein.

**(e)** Warten Sie zwischen den Messungen jeweils **1 Sekunde** (`time.sleep(1)`).

**(f)** Zeigen Sie am Ende das fertige Diagramm mit `plt.show()` an.

```python
import random
import time
import matplotlib.pyplot as plt

# === Aufgabe (a)-(f): Stresstest mit 10 Messungen ===

# Diagramm vorbereiten
plt.figure(figsize=(10, 4))
plt.ylabel('Temperatur (°C)')
plt.xlabel('Uhrzeit')
plt.title('Stresstest - Temperaturverlauf')

# Zähler für die while-Schleife
i = 0

while i < 10:
    # (a) Zufälligen Temperaturwert erzeugen
    temperatur = random.randint(50, 90)

    # (b) Aktuelle Uhrzeit ermitteln
    uhrzeit = time.strftime("%H:%M:%S", time.localtime())

    # (c) Messwert mit Uhrzeit ausgeben
    print(f"[{uhrzeit}] Messung {i + 1}: {temperatur} °C")

    # (d) Punkt im Diagramm eintragen
    plt.plot(uhrzeit, temperatur, 'r+')

    # (e) 1 Sekunde warten
    time.sleep(1)

    # Zähler erhöhen
    i = i + 1

# (f) Fertiges Diagramm anzeigen
plt.grid(True)
plt.show()
```
