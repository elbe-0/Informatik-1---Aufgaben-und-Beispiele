## Beispiele

Bevor Sie die Aufgabe lösen, machen Sie sich mit den folgenden Konzepten vertraut.
Führen Sie die Code-Zellen aus und beobachten Sie die Ausgabe.

```python
# Vergleich mit == in einer Schleife
teile = ["Kolben", "-", "Dichtung", "-", "Zündkerze"]

for i in range(len(teile)):
    if teile[i] != "-":
        print(f"Position {i}: {teile[i]}")
```

```python
# Eine 2D-Liste (Raster) mit verschachtelten Schleifen durchlaufen
lager_klein = [
    ["Kolben", "-", "Dichtung"],
    ["-", "Ventil", "-"],
    ["Zündkerze", "-", "-"]
]

for reihe_idx, reihe in enumerate(lager_klein):
    for spalte_idx, element in enumerate(reihe):
        if element != "-":
            print(f"Reihe {reihe_idx + 1}, Spalte {spalte_idx + 1}: {element}")
```
