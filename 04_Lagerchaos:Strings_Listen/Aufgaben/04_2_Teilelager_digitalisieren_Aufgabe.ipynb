## Beispiele

Die folgenden Beispiele zeigen die wichtigsten Listen-Operationen im Fabrikkontext.

````python
# --- Listen-Grundlagen im Fabrikkontext ---

# Liste erstellen
maschinen = ["Presse", "Fräse", "Drehbank"]
print("Maschinen:", maschinen)

# Auf einzelne Elemente zugreifen (Index beginnt bei 0)
print("Erste Maschine:", maschinen[0])   # Presse
print("Letzte Maschine:", maschinen[-1])  # Drehbank

# Länge einer Liste
print("Anzahl:", len(maschinen))  # 3

# Iteration mit for-Schleife
print("\nSchichtplan:")
schichten = ["Frühschicht", "Spätschicht", "Nachtschicht"]
for schicht in schichten:
    print("-", schicht)

# Element anhängen mit append()
maschinen.append("Schweißroboter")
print("\nNach append:", maschinen)

# Nummerierte Ausgabe mit enumerate()
print("\nMaschinenliste:")
for i, m in enumerate(maschinen, start=1):
    print(f"{i}. {m}")
````
