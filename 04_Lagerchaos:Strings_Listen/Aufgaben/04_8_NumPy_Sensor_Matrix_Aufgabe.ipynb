## Beispiele

Die folgenden Beispiele zeigen die wichtigsten NumPy-Grundlagen für die Arbeit mit Matrizen.

```python
# Beispiel 1: NumPy importieren und Arrays erstellen
import numpy as np

# Array aus einer Liste erstellen
temperaturen = np.array([19.5, 20.1, 21.3, 18.7, 20.0])
print("Temperaturen:", temperaturen)
print("Typ:", type(temperaturen))
```

```python
# Beispiel 2: Spezielle Arrays erstellen
import numpy as np

# 3x3-Matrix mit Nullen
nullen = np.zeros((3, 3))
print("Nullen-Matrix:")
print(nullen)

print()

# 3x3-Matrix gefüllt mit dem Wert 5
fuenfer = np.full((3, 3), 5)
print("Gefüllt mit 5:")
print(fuenfer)
```

```python
# Beispiel 3: Elementzugriff, Zeilen- und Spaltenzugriff
import numpy as np

matrix = np.array([
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
])

# Einzelnes Element: Zeile 0, Spalte 1
print("Element [0, 1]:", matrix[0, 1])

# Ganze Zeile 1
print("Zeile 1:", matrix[1, :])

# Ganze Spalte 2
print("Spalte 2:", matrix[:, 2])
```

```python
# Beispiel 4: Arithmetik und Broadcasting
import numpy as np

werte = np.array([10.0, 20.0, 30.0])

# Alle Werte verdoppeln (Broadcasting)
print("Verdoppelt:", werte * 2)

# Zwei Arrays subtrahieren (elementweise)
soll = np.array([20.0, 20.0, 20.0])
ist = np.array([19.5, 21.3, 20.0])
differenz = ist - soll
print("Differenz (ist - soll):", differenz)
```
