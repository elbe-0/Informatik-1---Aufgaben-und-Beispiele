## Beispiele

Die folgenden Beispiele zeigen nützliche Techniken zum Suchen und Ersetzen in Listen.

```python
# index(): findet die Position eines Elements in der Liste
teile = ["Schraube", "Mutter", "Bolzen", "Scheibe"]
position = teile.index("Bolzen")
print("Bolzen ist an Position:", position)
```

```python
# in-Operator: prüft, ob ein Element in der Liste enthalten ist
lagerbestand = ["Motor", "Getriebe", "Achse"]
print("Motor im Lager:", "Motor" in lagerbestand)
print("Bremse im Lager:", "Bremse" in lagerbestand)
```

```python
# Einfache Funktion: mit def definieren und aufrufen
def begruessung(name):
    print("Willkommen im Lager,", name)

begruessung("Herr Brandt")
```

```python
# Funktion mit Rückgabewert: Element in einer Liste suchen
def ist_vorhanden(lagerliste, teil):
    if teil in lagerliste:
        return True
    else:
        return False

teile = ["Kolben", "Dichtung", "Zündkerze"]
print("Kolben vorhanden?", ist_vorhanden(teile, "Kolben"))
print("Motor vorhanden?", ist_vorhanden(teile, "Motor"))
```
