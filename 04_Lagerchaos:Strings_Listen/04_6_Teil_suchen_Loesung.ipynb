## Beispiele

Bevor Sie die Aufgabe lösen, machen Sie sich mit den folgenden Konzepten vertraut.
Führen Sie die Code-Zellen aus und beobachten Sie die Ausgabe.

```python
# Vergleich mit == in einer Schleife
teile = ["Kolben", "---", "Dichtung", "---", "Zündkerze"]

for i in range(len(teile)):
    if teile[i] != "---":
        print(f"Position {i}: {teile[i]}")
```

```python
# Alternative: enumerate() liefert Index und Wert gleichzeitig
for reihe_idx, reihe in enumerate(teile):
    for spalte_idx, element in enumerate(reihe):
        print(f"Reihe {reihe_idx + 1}, Spalte {spalte_idx + 1}: {element}")
```

```python
# Vergleich mit == innerhalb verschachtelter Schleifen
gesucht = "Ventil"
for i in range(len(teile)):
    for j in range(len(teile[i])):
        if teile[i][j] == gesucht:
            print(f"{gesucht} gefunden in Reihe {i + 1}, Spalte {j + 1}!")
```

## Aufgabe

Erstellen Sie eine Suchfunktion für das Lagerraster der MotoTec-Fabrik.

**(a)** Verwenden Sie das unten vorbereitete Lagerraster. Geben Sie es zunächst mit
verschachtelten `for`-Schleifen aus.

**(b)** Fragen Sie den Benutzer mit `input()`, welches Teil gesucht werden soll.

**(c)** Durchsuchen Sie das gesamte Raster mit verschachtelten `for`-Schleifen.
Wenn das Teil gefunden wird, geben Sie die Position aus (Reihe und Spalte,
jeweils 1-indiziert für den Benutzer).

**(d)** Wenn das Teil nicht gefunden wird, geben Sie eine entsprechende Meldung aus.

```python
# Vorbereitetes Lagerraster (nicht verändern)
lager = [["---" for s in range(8)] for r in range(8)]
lager[0][2] = "Kolben"
lager[1][5] = "Dichtung"
lager[3][3] = "Turbolader"
lager[4][7] = "Zündkerze"
lager[6][1] = "Keilriemen"
```

```python
# === Aufgabe (a): Lagerraster ausgeben ===
print("Aktueller Lagerplan:")
print()
for reihe in lager:
    for element in reihe:
        print(element, end="\t")
    print()

# === Aufgabe (b): Benutzer nach dem gesuchten Teil fragen ===
suchbegriff = input("Welches Teil suchen Sie? ")

# === Aufgabe (c): Verschachtelte for-Schleifen - Raster durchsuchen ===
gefunden = False
for i in range(len(lager)):
    for j in range(len(lager[i])):
        if lager[i][j] == suchbegriff:
            print(f"{suchbegriff} gefunden in Reihe {i + 1}, Spalte {j + 1}!")
            gefunden = True

# === Aufgabe (d): Meldung wenn Teil nicht gefunden ===
if not gefunden:
    print(f"{suchbegriff} wurde nicht im Lager gefunden.")
```
