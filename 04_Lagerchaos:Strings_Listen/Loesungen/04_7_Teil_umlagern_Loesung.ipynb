## Briefing

In der MotoTec-Fabrik ist das Lager als zweidimensionales Raster organisiert.
Jede Position im Raster kann entweder ein Teil enthalten oder leer sein (dargestellt als `"---"`).
Im laufenden Betrieb müssen Teile regelmäßig umgelagert werden, zum Beispiel
um Wege zur Montagelinie zu verkürzen oder Platz für neue Lieferungen zu schaffen.
Ihre Aufgabe ist es, ein Programm zu schreiben, das ein Teil im Lager sucht,
von seiner alten Position entfernt und an einer neuen Position einlagert.

## Beispiele

Die folgenden Beispiele zeigen, wie Sie auf Elemente in einer 2D-Liste zugreifen
und diese verändern können.

```python
# Beispiel 1: Zugriff auf ein Element in einer 2D-Liste
regal = [
    ["Schraube", "Mutter", "---"],
    ["---", "Bolzen", "Dichtung"],
    ["Feder", "---", "Scheibe"]
]

# Zeile 1, Spalte 2 (0-basiert) => "Dichtung"
print("Element bei [1][2]:", regal[1][2])
```

```python
# Beispiel 2: Element verändern und auf Standardwert setzen
regal = [
    ["Schraube", "Mutter", "---"],
    ["---", "Bolzen", "Dichtung"],
    ["Feder", "---", "Scheibe"]
]

# "Bolzen" entfernen (Position leeren)
regal[1][1] = "---"
print("Nach Entfernen von Bolzen:")
for zeile in regal:
    print(zeile)
```

```python
# Beispiel 3: Neue Koordinaten einlesen und Teil platzieren
regal = [
    ["---", "---", "---"],
    ["---", "---", "---"],
    ["---", "---", "---"]
]

# Neue Position einlesen (1-basiert für den Benutzer)
neue_zeile = int(input("Neue Zeile (1-3): "))
neue_spalte = int(input("Neue Spalte (1-3): "))

# Teil an neuer Position einlagern (Umrechnung auf 0-basiert)
regal[neue_zeile - 1][neue_spalte - 1] = "Bolzen"

print("\nLager nach Einlagerung:")
for zeile in regal:
    print(zeile)
```

## Aufgabe

Schreiben Sie ein Programm, das ein Teil im Lager sucht, es von seiner aktuellen
Position entfernt und an einer neuen Position einlagert.

**Teilaufgaben:**

- **(a)** Starten Sie mit dem vorgegebenen 8x8-Lagerraster (siehe Code-Zelle unten). Geben Sie das Lager zeilenweise aus.
- **(b)** Fragen Sie per `input()`, welches Teil umgelagert werden soll. Suchen Sie es mit verschachtelten Schleifen. Falls das Teil nicht gefunden wird, geben Sie eine Fehlermeldung aus.
- **(c)** Leeren Sie die alte Position, indem Sie den Wert auf `"---"` setzen.
- **(d)** Fragen Sie nach der neuen Zeile und Spalte (1-8). Platzieren Sie das Teil an der neuen Position.
- **(e)** Geben Sie das aktualisierte Lager zeilenweise aus.

```python
# === Vorgegebenes Lager (nicht verändern) ===
lager = [["---" for s in range(8)] for r in range(8)]
lager[0][2] = "Kolben"
lager[1][5] = "Dichtung"
lager[3][3] = "Turbolader"
lager[4][7] = "Zündkerze"
lager[6][1] = "Keilriemen"

# === Aufgabe (a): Lager ausgeben ===
print("Aktueller Lagerbestand:")
for zeile in lager:
    print(zeile)

# === Aufgabe (b): Teil suchen ===
teil_name = input("\nWelches Teil soll umgelagert werden? ")

gefunden = False
gefunden_zeile = -1
gefunden_spalte = -1

for i in range(len(lager)):
    for j in range(len(lager[i])):
        if lager[i][j] == teil_name:
            gefunden_zeile = i
            gefunden_spalte = j
            gefunden = True

if not gefunden:
    print("Fehler: Das Teil", teil_name, "wurde nicht im Lager gefunden.")
else:
    print(teil_name, "gefunden bei Zeile", gefunden_zeile + 1, ", Spalte", gefunden_spalte + 1)

    # === Aufgabe (c): Alte Position leeren ===
    lager[gefunden_zeile][gefunden_spalte] = "---"

    # === Aufgabe (d): Neue Position einlesen und Teil platzieren ===
    neue_zeile = int(input("Neue Zeile (1-8): "))
    neue_spalte = int(input("Neue Spalte (1-8): "))

    lager[neue_zeile - 1][neue_spalte - 1] = teil_name
    print(teil_name, "wurde nach Zeile", neue_zeile, ", Spalte", neue_spalte, "umgelagert.")

    # === Aufgabe (e): Aktualisiertes Lager ausgeben ===
    print("\nAktualisierter Lagerbestand:")
    for zeile in lager:
        print(zeile)
```

## Erklärung

Das Programm durchläuft das gesamte 8x8-Lagerraster mit zwei verschachtelten
`for`-Schleifen. Sobald das gesuchte Teil gefunden wird, werden Zeile und Spalte
gespeichert. Anschließend wird die alte Position auf `"---"` gesetzt (geleert)
und das Teil an der vom Benutzer eingegebenen neuen Position platziert.

Beachten Sie die Umrechnung zwischen 1-basierten Benutzereingaben und 0-basierten
Python-Indizes: `lager[zeile - 1][spalte - 1]`.
