## Beispiele

Die folgenden Beispiele zeigen nützliche Techniken zum Suchen und Ersetzen in Listen.

```python
# index(): findet die Position eines Elements in der Liste
teile = ["Schraube", "Mutter", "Bolzen", "Scheibe"]
position = teile.index("Bolzen")
print("Bolzen ist an Position:", position)
```

```python
# in-Operator: prüft, ob ein Element in der Liste enthalten ist
lagerbestand = ["Motor", "Getriebe", "Achse"]
print("Motor im Lager:", "Motor" in lagerbestand)
print("Bremse im Lager:", "Bremse" in lagerbestand)
```

```python
# Einfache Funktion: mit def definieren und aufrufen
def begruessung(name):
    print("Willkommen im Lager,", name)

begruessung("Herr Brandt")
```

```python
# Funktion mit Rückgabewert: Element in einer Liste suchen
def ist_vorhanden(lagerliste, teil):
    if teil in lagerliste:
        return True
    else:
        return False

teile = ["Kolben", "Dichtung", "Zündkerze"]
print("Kolben vorhanden?", ist_vorhanden(teile, "Kolben"))
print("Motor vorhanden?", ist_vorhanden(teile, "Motor"))
```

## Aufgabe

Ausgangsliste: `lager = ["Kolben", "Dichtung", "Zündkerze", "Keilriemen"]`

**(a)** Schreiben Sie eine Funktion `suchen_ersetzen(liste, alt, neu)`, die das Element
`alt` in `liste` sucht (mit `index()`) und durch `neu` ersetzt.

**(b)** Fragen Sie den Benutzer per `input()`, welches Teil ersetzt werden soll und wie
das neue Teil heißt. Rufen Sie dann `suchen_ersetzen()` auf.

**(c)** Geben Sie die aktualisierte Lagerliste aus.

```python
# === Aufgabe (a): Funktion suchen_ersetzen() schreiben ===
def suchen_ersetzen(liste, alt, neu):
    index = liste.index(alt)
    liste[index] = neu
```

```python
# === Aufgabe (b): Benutzereingabe und Funktionsaufruf ===
lager = ["Kolben", "Dichtung", "Zündkerze", "Keilriemen"]
teil_alt = input("Welches Teil soll ersetzt werden? ")
teil_neu = input("Wie heißt das neue Teil? ")
suchen_ersetzen(lager, teil_alt, teil_neu)
```

```python
# === Aufgabe (c): Aktualisierte Liste ausgeben ===
print("Aktualisiertes Lager:", lager)
```
