## Beispiele

Bevor Sie die Aufgabe lösen, machen Sie sich mit den folgenden Konzepten vertraut.
Führen Sie die Code-Zellen aus und beobachten Sie die Ausgabe.

```python
# 2D-Liste mit List-Comprehension erstellen
# Eine 3x3-Matrix, gefüllt mit "-"
matrix = [["-" for spalte in range(3)] for reihe in range(3)]
print("Leere 3x3-Matrix:")
print(matrix)
```

```python
# Auf einzelne Elemente zugreifen: matrix[reihe][spalte]
matrix[0][1] = "Kolben"
matrix[2][2] = "Dichtung"
print("Element in Reihe 1, Spalte 2:", matrix[0][1])
print("Element in Reihe 3, Spalte 3:", matrix[2][2])
print("Gesamte Matrix:", matrix)
```

```python
# Verschachtelte for-Schleifen zur Anzeige eines Rasters
for reihe in matrix:
    for element in reihe:
        print(element, end="\t")
    print()  # Zeilenumbruch nach jeder Reihe
```

```python
# clear_output() löscht die bisherige Ausgabe der Zelle
from IPython.display import clear_output

# Beispiel: clear_output() wird typischerweise in while-Schleifen verwendet,
# um die Anzeige vor jeder neuen Ausgabe zu bereinigen.
print("Diese Ausgabe bleibt sichtbar.")
# clear_output()  # Entkommentieren Sie diese Zeile, um den Effekt zu sehen
```

## Aufgabe

Schreiben Sie ein Programm, das dem Benutzer das Einlagern von beliebig vielen Teilen
in ein 8x8-Lagerraster ermöglicht. Jeder Platz soll initial mit `"-"` belegt sein.
Fragen Sie den Benutzer nach Teilname, Reihe und Spalte, lagern Sie das Teil ein und
geben Sie die Belegung des Lagers nach jedem Durchlauf aus.

```python
from IPython.display import clear_output

# 8x8-Raster erstellen
lager = [["-" for spalte in range(8)] for reihe in range(8)]

# Teile interaktiv einlagern
weitermachen = True
while weitermachen:
    teil = input("Welches Teil einlagern? ")
    reihe = int(input("Reihe (1-8): "))
    spalte = int(input("Spalte (1-8): "))
    lager[reihe - 1][spalte - 1] = teil

    clear_output()
    print("Aktueller Lagerplan:\n")
    for r in lager:
        for element in r:
            print(element, end="\t")
        print()

    antwort = input("Weiteres Teil einlagern? (j/n): ")
    if antwort == "n":
        weitermachen = False

# Finalen Lagerplan ausgeben
clear_output()
print("Finaler Lagerplan:\n")
for reihe in lager:
    for element in reihe:
        print(element, end="\t")
    print()
```
