## Beispiele

Die folgenden Beispiele zeigen die wichtigsten String-Operationen im Fabrikkontext.

````python
# --- String-Grundlagen im Fabrikkontext ---

# String-Verkettung (Concatenation)
halle = "Halle"
nummer = "7"
standort = halle + "-" + nummer
print("Standort:", standort)  # Halle-7

# Länge mit len()
maschinen_id = "PRESS-042-A"
print("Länge der ID:", len(maschinen_id))  # 11

# Indexing: einzelne Zeichen lesen
print("Erstes Zeichen:", maschinen_id[0])    # P
print("Letztes Zeichen:", maschinen_id[-1])   # A

# Slicing: Teilstrings extrahieren
typ = maschinen_id[0:5]    # "PRESS"
code = maschinen_id[6:9]   # "042"
print("Typ:", typ, "| Code:", code)

# find(): Position eines Teilstrings
beschreibung = "Wartung Pressmaschine Halle-3"
pos = beschreibung.find("Halle")
print("'Halle' beginnt an Position:", pos)  # 22

# replace(): Teilstring ersetzen
fehlerhaft = "Pruefprotokoll Oelwechsel"
korrekt = fehlerhaft.replace("ue", "ü").replace("Oel", "Öl")
print("Korrigiert:", korrekt)  # Prüfprotokoll Ölwechsel
````

## Aufgabe

In der Zulieferer-Datenbank ist folgender fehlerhafter Eintrag gespeichert:

```python
eintrag = "GmbH, MotoTec, Zulieferer"
```

Der korrekte Firmenname lautet: **MotoTec Zulieferer GmbH**

**(a)** Ermitteln Sie mit `len()` die Länge des fehlerhaften Strings und geben Sie sie aus.

**(b)** Extrahieren Sie mit Slicing die drei Bestandteile `"GmbH"`, `"MotoTec"` und `"Zulieferer"` als einzelne Variablen. Nutzen Sie `find()`, um die Positionen der Kommas zu bestimmen.

**(c)** Setzen Sie den korrekten Namen `"MotoTec Zulieferer GmbH"` aus den Teilen zusammen (String-Verkettung) und geben Sie ihn aus.

**(d)** Ein zweiter Datenbank-Eintrag enthält falsch kodierte Umlaute:

```python
beschreibung = "Qualitaetspruefung Motoroelwechsel"
```

Korrigieren Sie mit `replace()` die Kodierungen: `ae` → `ä`, `oe` → `ö`, `ue` → `ü`. Geben Sie das Ergebnis aus.

````python
# Fehlerhafter Datenbank-Eintrag
eintrag = "GmbH, MotoTec, Zulieferer"

# === Aufgabe (a): Länge des Strings ermitteln ===
laenge = len(eintrag)
print("Länge des Eintrags:", laenge)

# === Aufgabe (b): Bestandteile per Slicing extrahieren ===
komma1 = eintrag.find(",")
komma2 = eintrag.find(",", komma1 + 1)

teil1 = eintrag[0:komma1]               # "GmbH"
teil2 = eintrag[komma1 + 2:komma2]      # "MotoTec"
teil3 = eintrag[komma2 + 2:]            # "Zulieferer"

print("Teil 1:", teil1)
print("Teil 2:", teil2)
print("Teil 3:", teil3)

# === Aufgabe (c): Korrekten Namen zusammensetzen ===
name = teil2 + " " + teil3 + " " + teil1
print("Korrekter Name:", name)

# === Aufgabe (d): Umlaute reparieren ===
beschreibung = "Qualitaetspruefung Motoroelwechsel"
beschreibung = beschreibung.replace("ae", "ä").replace("oe", "ö").replace("ue", "ü")
print("Korrigiert:", beschreibung)
````

### Erklärung

- `find(",")` liefert die Position des ersten Kommas (Index 4). Mit `find(",", 5)` suchen
  wir ab Position 5 weiter und finden das zweite Komma.
- Beim Slicing ist `eintrag[komma1 + 2:komma2]` nötig, weil nach jedem Komma ein
  Leerzeichen steht (`", "` = 2 Zeichen).
- `replace()` kann verkettet werden, da es immer einen neuen String zurückgibt.
