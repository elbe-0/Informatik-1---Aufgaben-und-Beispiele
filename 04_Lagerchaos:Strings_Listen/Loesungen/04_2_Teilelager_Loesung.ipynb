## Beispiele

Die folgenden Beispiele zeigen die wichtigsten Listen-Operationen im Fabrikkontext.

```python
teile = ["Kolben", "Dichtung", "Zündkerze", "Keilriemen"]
print(teile)
```

```python
# Zugriff auf ein Element über den Index
print("Das 3. Element hat den Wert", teile[2])
teile[2] = "Zündspule"
print("Das 3. Element hat den Wert", teile[2])
```

```python
# Zugriff auf alle Elemente mit einer for-Schleife
for teil in teile:
    print(teil)
```

#### Python stellt eine Reihe von Methoden zum Arbeiten mit Listen zur Verfügung:

```python
# Anfügen von Elementen
teile.append("Ölfilter")
for teil in teile:
    print(teil)
```

```python
# Einfügen von Elementen an einer bestimmten Stelle
teile.insert(1, "Bremsbelag")
for teil in teile:
    print(teil)
```

```python
# Löschen eines Elementes an einer bestimmten Stelle
teile.pop(3)
for teil in teile:
    print(teil)
```

```python
# Löschen eines Elementes mit einem bestimmten Wert
teile.remove("Ölfilter")
for teil in teile:
    print(teil)
```

## Aufgaben

### Aufgabe 4.2a

Gegeben sind vier Variablen mit Teilenamen. Überführen Sie die Variablen in eine Liste.
Geben Sie die Liste auf dem Bildschirm aus.

```python
teil1 = "Kolben"
teil2 = "Dichtung"
teil3 = "Zündkerze"
teil4 = "Keilriemen"

lager = [teil1, teil2, teil3, teil4]
for teil in lager:
    print(teil)
```

### Aufgabe 4.2b

Löschen Sie aus Ihrem Ergebnis das erste Element und ersetzen Sie das neue zweite Element
durch ein Teil Ihrer Wahl. Der Benutzer soll das neue Teil eingeben können.
Geben Sie das Ergebnis auf dem Bildschirm aus.

```python
lager.pop(0)
neues_teil = input("Neues Teil für Position 2: ")
lager[1] = neues_teil
for teil in lager:
    print(teil)
```

### Aufgabe 4.2c

Schreiben Sie ein Programm, das dem Benutzer das Auswählen und Ändern eines
Elements mit einem bestimmten Wert ermöglicht. Geben Sie das Ergebnis auf dem
Bildschirm aus.

```python
teil_alt = input("Welches Teil soll ersetzt werden? ")
teil_neu = input("Wie heißt das neue Teil? ")

position = lager.index(teil_alt)
lager[position] = teil_neu

for teil in lager:
    print(teil)
```
