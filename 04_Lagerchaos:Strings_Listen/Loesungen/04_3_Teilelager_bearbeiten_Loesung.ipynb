## Beispiele

Die folgenden Beispiele zeigen, wie Sie Listenelemente im Lagerkontext bearbeiten können.

```python
# pop() ohne Index: entfernt das letzte Element und gibt es zurück
werkzeuge = ["Hammer", "Zange", "Schraubenzieher"]
entfernt = werkzeuge.pop()
print("Entfernt:", entfernt)
print("Werkzeuge:", werkzeuge)
```

```python
# pop(index): entfernt das Element an der angegebenen Position
ersatzteile = ["Schraube", "Mutter", "Bolzen", "Scheibe"]
entfernt = ersatzteile.pop(0)
print("Entfernt:", entfernt)
print("Ersatzteile:", ersatzteile)
```

```python
# remove(): entfernt das erste Vorkommen eines bestimmten Wertes
materialien = ["Stahl", "Kupfer", "Aluminium", "Kupfer"]
materialien.remove("Kupfer")
print("Materialien:", materialien)
```

```python
# del: löscht ein Element per Index
bauteile = ["Motor", "Getriebe", "Achse"]
del bauteile[1]
print("Bauteile:", bauteile)
```

```python
# Index-Zuweisung: ein Element direkt ersetzen
lagerbestand = ["Schraube M6", "Schraube M8", "Schraube M10"]
lagerbestand[1] = "Schraube M12"
print("Lagerbestand:", lagerbestand)
```

```python
# input() mit Index-Zuweisung: Benutzer gibt den neuen Wert ein
teile = ["Ventil", "Feder", "Kolbenring"]
neues_teil = input("Neues Teil für Position 0: ")
teile[0] = neues_teil
print("Teile:", teile)
```

## Aufgabe

Ausgangsliste: `lager = ["Kolben", "Dichtung", "Zündkerze", "Keilriemen", "Ölfilter"]`

**(a)** Der Kolben wurde verbaut und muss aus dem Lager entfernt werden.
Entfernen Sie den Kolben aus dem Lager. Geben Sie die aktualisierte Liste aus.

**(b)** Das zweite Element in der aktualisierten Liste ist falsch erfasst.
Lesen Sie per `input()` einen neuen Teilenamen ein und ersetzen Sie das Element
an Index `1`. Geben Sie die aktualisierte Liste aus.

**(c)** Der Ölfilter wird nicht mehr benötigt. Entfernen Sie `"Ölfilter"` mit `remove()`.
Geben Sie die finale Liste aus.

```python
# === Aufgabe (a): Kolben entfernen ===
lager = ["Kolben", "Dichtung", "Zündkerze", "Keilriemen", "Ölfilter"]
lager.pop(0)
print("Lager nach Entfernung:", lager)
```

```python
# === Aufgabe (b): Zweites Element per input() ersetzen ===
neuer_name = input("Bitte geben Sie den korrekten Teilenamen ein: ")
lager[1] = neuer_name
print("Lager nach Korrektur:", lager)
```

```python
# === Aufgabe (c): Ölfilter mit remove() entfernen ===
lager.remove("Ölfilter")
print("Finale Lagerliste:", lager)
```
