## Beispiele

Die folgenden Beispiele zeigen die wichtigsten NumPy-Grundlagen für die Arbeit mit Matrizen.

```python
# Beispiel 1: NumPy importieren und Arrays erstellen
import numpy as np

# Array aus einer Liste erstellen
temperaturen = np.array([19.5, 20.1, 21.3, 18.7, 20.0])
print("Temperaturen:", temperaturen)
print("Typ:", type(temperaturen))
```

```python
# Beispiel 2: Spezielle Arrays erstellen
import numpy as np

# 3x3-Matrix mit Nullen
nullen = np.zeros((3, 3))
print("Nullen-Matrix:")
print(nullen)

print()

# 3x3-Matrix gefüllt mit dem Wert 5
fuenfer = np.full((3, 3), 5)
print("Gefüllt mit 5:")
print(fuenfer)
```

```python
# Beispiel 3: Elementzugriff, Zeilen- und Spaltenzugriff
import numpy as np

matrix = np.array([
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
])

# Einzelnes Element: Zeile 0, Spalte 1
print("Element [0, 1]:", matrix[0, 1])

# Ganze Zeile 1
print("Zeile 1:", matrix[1, :])

# Ganze Spalte 2
print("Spalte 2:", matrix[:, 2])
```

```python
# Beispiel 4: Arithmetik und Broadcasting
import numpy as np

werte = np.array([10.0, 20.0, 30.0])

# Alle Werte verdoppeln (Broadcasting)
print("Verdoppelt:", werte * 2)

# Zwei Arrays subtrahieren (elementweise)
soll = np.array([20.0, 20.0, 20.0])
ist = np.array([19.5, 21.3, 20.0])
differenz = ist - soll
print("Differenz (ist - soll):", differenz)
```

## Aufgabe

Arbeiten Sie mit NumPy-Arrays, um die Temperatursensoren der Fabrik auszuwerten.

**Teilaufgaben:**

- **(a)** Erstellen Sie ein 8x8 NumPy-Array `soll_temperatur` mit `np.full()`, initialisiert auf `20.0` (Soll-Temperatur aller Maschinen in °C). Geben Sie es aus.
- **(b)** Maschinenreihe 3 (Index 2) überhitzt! Setzen Sie die gesamte Zeile auf `85.0` mithilfe von Zeilen-Zugriff (Broadcasting). Geben Sie das Array aus.
- **(c)** Erstellen Sie ein zweites Array `ist_temperatur` mit den vorgegebenen Messwerten (siehe Code-Zelle). Berechnen Sie die Differenz `differenz = ist_temperatur - soll_temperatur`. Geben Sie die Differenz-Matrix aus.
- **(d)** Erstellen Sie `backup = soll_temperatur` (ohne `.copy()`). Ändern Sie `backup[0, 0] = 999`. Geben Sie beide Arrays aus und beobachten Sie, dass sich `soll_temperatur` ebenfalls verändert hat! Führen Sie dann die korrekte Kopie mit `.copy()` durch.

```python
import numpy as np

# === Aufgabe (a): Soll-Temperatur-Array erstellen ===
soll_temperatur = np.full((8, 8), 20.0)
print("Soll-Temperaturen (alle Maschinen):")
print(soll_temperatur)
```

```python
# === Aufgabe (b): Zeile 3 (Index 2) auf 85.0 setzen ===
soll_temperatur[2, :] = 85.0
print("Soll-Temperaturen nach Überhitzung in Reihe 3:")
print(soll_temperatur)
```

```python
# === Aufgabe (c): Ist-Temperaturen und Differenz berechnen ===
ist_temperatur = np.array([
    [19.8, 20.3, 20.1, 19.7, 20.5, 19.9, 20.2, 20.0],
    [20.1, 19.6, 20.4, 20.0, 19.8, 20.3, 19.7, 20.2],
    [82.5, 84.1, 86.3, 81.9, 85.7, 83.2, 87.0, 84.8],
    [20.3, 19.9, 20.0, 20.2, 19.5, 20.1, 20.4, 19.8],
    [19.7, 20.2, 20.1, 19.8, 20.0, 20.3, 19.6, 20.4],
    [20.0, 19.8, 20.5, 20.1, 19.9, 20.2, 20.0, 19.7],
    [20.2, 20.0, 19.9, 20.4, 20.1, 19.6, 20.3, 20.0],
    [19.9, 20.1, 20.3, 19.8, 20.0, 20.4, 19.7, 20.2]
])

differenz = ist_temperatur - soll_temperatur
print("Differenz (Ist - Soll):")
print(differenz)
```

## Erklärung zur Differenz

In der Differenz-Matrix sehen Sie, dass die normalen Maschinenreihen nur geringe
Abweichungen zeigen (im Bereich von -0.5 bis +0.5 °C). Die überhitzte Reihe 3
zeigt negative Werte (z.B. -2.5, -0.9), weil die Ist-Temperatur dort unter dem
angepassten Sollwert von 85.0 °C liegt. Die elementweise Subtraktion von NumPy
berechnet alle 64 Differenzen in einer einzigen Operation.

```python
# === Aufgabe (d): Copy-Problem demonstrieren ===

# Falsch: Nur eine Referenz (keine echte Kopie)
backup = soll_temperatur
backup[0, 0] = 999

print("backup[0, 0]:", backup[0, 0])
print("soll_temperatur[0, 0]:", soll_temperatur[0, 0])
print("Beide zeigen 999 - es ist dasselbe Objekt!")

# Reparieren: Wert zurücksetzen
soll_temperatur[0, 0] = 20.0

print()

# Richtig: Echte Kopie mit .copy()
backup = soll_temperatur.copy()
backup[0, 0] = 999

print("backup[0, 0]:", backup[0, 0])
print("soll_temperatur[0, 0]:", soll_temperatur[0, 0])
print("Jetzt sind es zwei unabhängige Arrays!")
```

## Erklärung zu .copy()

In Python erzeugt eine einfache Zuweisung wie `backup = soll_temperatur` keine Kopie.
Stattdessen verweisen beide Variablen auf dasselbe Objekt im Speicher.
Änderungen über eine Variable wirken sich daher auf die andere aus.
Mit `backup = soll_temperatur.copy()` wird eine echte, unabhängige Kopie erstellt.
Dieses Verhalten gilt sowohl für NumPy-Arrays als auch für Python-Listen.
