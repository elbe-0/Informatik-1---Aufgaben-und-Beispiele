## Briefing

Das Terminal läuft, aber die Fabrik liegt noch im Dunkeln. Bevor weitere Systeme hochgefahren werden können, muss die Stromversorgung geprüft werden. Zwei unabhängige Stromkreise versorgen die Fabrikhalle - deren Spannungswerte müssen addiert werden, um die Gesamtkapazität zu berechnen. Nur wenn die Gesamtspannung stimmt, kann das Team die nächsten Systeme zuschalten.

## Stakeholder-Dialog

*Sandra Holz (IT-Leiterin) zeigt auf die Anzeige der Stromverteilung. Ihre Stimme ist angespannt:*

> *"Die Spannungswerte müssen stimmen, sonst fliegt uns die Sicherung. Lesen Sie die beiden Werte ein und rechnen Sie die Gesamtspannung aus. Und prüfen Sie das Ergebnis - wenn da Unsinn rauskommt, haben wir ein Problem."*

## Beispiele

Zur Erinnerung — so funktioniert die Eingabe von Zahlen:

````python
spannung = input("Spannung eingeben: ")
print("Eingegebener Wert: " + spannung)
````

## Aufgabe

Schreiben Sie ein Python-Programm, das die Spannungswerte zweier Stromkreise einliest und die Gesamtspannung berechnet:

**(a)** Lesen Sie zwei Spannungswerte per `input()` ein und versuchen Sie, diese mit `+` zu addieren. Geben Sie das Ergebnis aus.

Probieren Sie z.B. die Werte `230` und `120` aus. Was fällt Ihnen am Ergebnis auf?

**(b)** Erklären Sie das Problem: Warum zeigt das Programm `230120` statt `350`?

**(c)** Beheben Sie das Problem. Geben Sie die korrekte Gesamtspannung aus, z.B.:

```
Gesamtspannung: 350 Volt
```

````python
# (a) Zwei Spannungswerte einlesen und mit + "addieren":
spannung1 = input("Spannung Stromkreis 1 eingeben: ")
spannung2 = input("Spannung Stromkreis 2 eingeben: ")
ergebnis = spannung1 + spannung2
print("Ergebnis:", ergebnis)

# (b) Was passiert? Warum ist das Ergebnis falsch?
# input() gibt immer einen String zurück.
# Der +-Operator verkettet bei Strings: "230" + "120" ergibt "230120".

# (c) Problem beheben und korrekt addieren:
spannung1 = int(input("Spannung Stromkreis 1 eingeben: "))
spannung2 = int(input("Spannung Stromkreis 2 eingeben: "))
gesamt = spannung1 + spannung2
print("Gesamtspannung:", gesamt, "Volt")
````
