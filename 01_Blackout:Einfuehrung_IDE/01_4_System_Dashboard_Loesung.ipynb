**Thema:** Formatierung mit `\t` und `\n`

## Briefing

Die Bootsequenz läuft, die Stromversorgung ist geprüft, und die Statusanzeige meldet erste Ergebnisse. Jetzt braucht das Team den entscheidenden Gesamtüberblick: Welche Systeme laufen, welche nicht? Sandra Holz will ein zentrales Dashboard, das alle kritischen Fabriksysteme auf einen Blick zeigt. Erst wenn dieses Dashboard steht, kann das Team die nächsten Schritte planen.

## Stakeholder-Dialog

*Sandra Holz (IT-Leiterin) tritt an das Terminal. Ihr Blick ist konzentriert:*

> *"Ich brauch eine Übersicht aller Systeme. Auf einen Blick. Strom, Wasser, Netzwerk, Produktion - was läuft, was nicht. Und zwar ordentlich formatiert, nicht irgendein Chaos auf dem Bildschirm. Ein einziger Befehl, ein sauberes Dashboard. Kriegen Sie das hin?"*

## Aufgabe

Geben Sie mit einem **einzigen** `print()`-Aufruf eine kurze Systemübersicht aus. Die Ausgabe soll so aussehen:

```
MOTOTEC SYSTEM-STATUS
Strom:          OK
Wasser:         FEHLER
Netzwerk:       OK
```

Verwenden Sie:
- `\n` für die **Zeilenumbrüche** zwischen den Zeilen
- `\t` für die **Abstände** zwischen Systemname und Status
- Alles in **einem einzigen** `print()`-Aufruf

---

**Hinweise:**
- `\n` erzeugt einen Zeilenumbruch: `print("Zeile 1\nZeile 2")`
- `\t` erzeugt einen Tabulator-Abstand: `print("Name:\tWert")`
- Sie können `\t` und `\n` beliebig kombinieren

````python
# === Lösung: System-Dashboard ===
# Mit \n erzeugen wir Zeilenumbrüche, mit \t Tabulator-Abstände.
# Alles in einem einzigen print()-Aufruf.
print("MOTOTEC SYSTEM-STATUS\nStrom:\t\tOK\nWasser:\t\tFEHLER\nNetzwerk:\tOK")
````
