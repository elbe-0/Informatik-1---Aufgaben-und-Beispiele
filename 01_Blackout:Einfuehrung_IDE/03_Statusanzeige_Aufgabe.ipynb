````python
# (c) Ladebalken für den Systemstart
# Die folgenden Zeilen sind vorgegeben - ändern Sie nur die print()-Zeile!

import time

print("Systemstart wird eingeleitet...")

fortschritt = 0
while fortschritt <= 100:
    balken = "#" * (fortschritt // 5)
    luecke = "." * (20 - fortschritt // 5)

    # IHRE AUFGABE: Passen Sie die folgende print()-Zeile an.
    # Verwenden Sie \r am Anfang des Strings und end="",
    # damit der Balken sich auf derselben Zeile aktualisiert.
    print("[" + balken + luecke + "] " + str(fortschritt) + "%")

    time.sleep(0.1)
    fortschritt = fortschritt + 5

print("\nAlle Systeme bereit.")
````
