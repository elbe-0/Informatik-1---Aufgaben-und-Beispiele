# Aufgabe 1.2: Stromversorgung (Lösung)

**Modul:** Informatik I - Mission MotoTec  
**Thema:** `input()`, `int()`, String vs. Integer

## Briefing

Das Terminal läuft, aber die Fabrik liegt noch im Dunkeln. Bevor weitere Systeme hochgefahren werden können, muss die Stromversorgung geprüft werden. Zwei unabhängige Stromkreise versorgen die Fabrikhalle - deren Spannungswerte müssen addiert werden, um die Gesamtkapazität zu berechnen. Nur wenn die Gesamtspannung stimmt, kann das Team die nächsten Systeme zuschalten.

## Stakeholder-Dialog

*Sandra Holz (IT-Leiterin) zeigt auf die Anzeige der Stromverteilung. Ihre Stimme ist angespannt:*

> *"Die Spannungswerte müssen stimmen, sonst fliegt uns die Sicherung. Lesen Sie die beiden Werte ein und rechnen Sie die Gesamtspannung aus. Und prüfen Sie das Ergebnis - wenn da Unsinn rauskommt, haben wir ein Problem."*

## Aufgabe

Schreiben Sie ein Python-Programm, das die Spannungswerte zweier Stromkreise einliest und die Gesamtspannung berechnet:

**(a)** Lesen Sie zwei Spannungswerte per `input()` ein und versuchen Sie, diese mit `+` zu addieren. Geben Sie das Ergebnis aus.

Probieren Sie z.B. die Werte `230` und `120` aus. Was fällt Ihnen am Ergebnis auf?

**(b)** Erklären Sie das Problem: Warum zeigt das Programm `230120` statt `350`?

**(c)** Beheben Sie das Problem, indem Sie die Eingaben mit `int()` in Ganzzahlen umwandeln, bevor Sie sie addieren. Geben Sie die korrekte Gesamtspannung aus, z.B.:

```
Gesamtspannung: 350 Volt
```

---

**Hinweise:**
- `input()` gibt immer einen **String** (Text) zurück - auch wenn eine Zahl eingegeben wird
- Bei Strings bewirkt `+` eine **Verkettung**: `"230" + "120"` ergibt `"230120"`
- `int("230")` wandelt den String `"230"` in die Ganzzahl `230` um

````python
# === Aufgabe (a): Spannungswerte einlesen und "addieren" ===
# input() gibt IMMER einen String zurück - auch wenn man eine Zahl eingibt.
# Das bedeutet: spannung1 und spannung2 sind Strings, keine Zahlen!
spannung1 = input("Spannung Stromkreis 1 (Volt): ")
spannung2 = input("Spannung Stromkreis 2 (Volt): ")

# Bei Strings bewirkt + eine Verkettung (Konkatenation), keine Addition.
# "230" + "120" ergibt "230120" statt 350!
ergebnis_falsch = spannung1 + spannung2
print("Ergebnis mit Strings: " + ergebnis_falsch)

# === Aufgabe (b): Das Problem ===
# input() liefert immer Text (String). Python kann nicht wissen,
# ob "230" eine Zahl oder ein Wort sein soll. Deshalb behandelt
# der +-Operator die Eingaben als Text und hängt sie aneinander.

# === Aufgabe (c): Mit int() korrekt addieren ===
# int() wandelt einen String in eine Ganzzahl (Integer) um.
# Erst nach der Umwandlung rechnet Python mit echten Zahlen.
zahl1 = int(spannung1)
zahl2 = int(spannung2)

gesamt = zahl1 + zahl2
print("Gesamtspannung: " + str(gesamt) + " Volt")
````
