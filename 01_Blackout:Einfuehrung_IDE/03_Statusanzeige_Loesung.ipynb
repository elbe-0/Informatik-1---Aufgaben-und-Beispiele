## Briefing

Das Terminal laeuft, die Stromversorgung ist geprueft - aber der Bildschirm zeigt nur wirren Text. Die Statusanzeige des Fabriksystems ist voellig unformatiert: Meldungen ohne Absaetze, Tabellen ohne Spalten, alles ein unlesbarer Block. Bevor das Team die einzelnen Systeme hochfahren kann, muss die Statusanzeige repariert werden - sonst weiss niemand, welches System laeuft und welches nicht.

## Aufgabe

Reparieren Sie die Statusanzeige des MotoTec-Terminals mit Hilfe von Steuerzeichen:

**(a)** Erstellen Sie ein formatiertes Status-Dashboard mit `\t` (Tabulator) und `\n` (Zeilenumbruch). Die Ausgabe soll so aussehen:

```
=== MOTOTEC SYSTEM-STATUS ===

System		Status		Prioritaet
-------		------		----------
Strom		ONLINE		HOCH
Netzwerk	OFFLINE		HOCH
Beleuchtung	OFFLINE		MITTEL
Heizung		OFFLINE		NIEDRIG
```

Verwenden Sie `\t` fuer die Spaltenausrichtung und `\n` fuer die Zeilenumbrueche. Sie koennen mehrere `print()`-Aufrufe verwenden oder alles in einem einzigen `print()` mit `\n` ausgeben.

**(b)** Testen Sie die Steuerzeichen `\b` (Backspace) und `\r` (Carriage Return). Fuehren Sie die folgenden Zeilen aus und beobachten Sie die Ausgabe:

```python
print("Status: FEHLERR\b ")
print("Ladevorgang...\rFERTIG")
```

Erklaeren Sie als Kommentar im Code: Was bewirkt `\b`? Was bewirkt `\r`?

**(c)** Bauen Sie einen Ladebalken fuer den Systemstart. Die Vorlage unten verwendet `import time` und eine `while`-Schleife - diese Werkzeuge werden spaeter im Kurs erklaert. Ihre Aufgabe ist es, die `print()`-Zeile so anzupassen, dass der Ladebalken sich **auf derselben Zeile aktualisiert** (mit `\r` und `end=""`).

---

**Hinweise:**
- `\n` erzeugt einen Zeilenumbruch (neue Zeile)
- `\t` erzeugt einen Tabulator (Einrueckung zur naechsten Tabstopp-Position)
- `\b` loescht das vorherige Zeichen (Backspace)
- `\r` setzt den Cursor an den Anfang der aktuellen Zeile zurueck (Carriage Return)
- `print("Text", end="")` unterdrueckt den automatischen Zeilenumbruch am Ende

````python
# === Aufgabe (a): Status-Dashboard mit \t und \n formatieren ===
# Mit \t (Tabulator) richten wir die Spalten aus.
# Mit \n (Zeilenumbruch) erzeugen wir neue Zeilen.

print("=== MOTOTEC SYSTEM-STATUS ===")
print()
print("System\t\tStatus\t\tPrioritaet")
print("-------\t\t------\t\t----------")
print("Strom\t\tONLINE\t\tHOCH")
print("Netzwerk\tOFFLINE\t\tHOCH")
print("Beleuchtung\tOFFLINE\t\tMITTEL")
print("Heizung\t\tOFFLINE\t\tNIEDRIG")
````

````python
# === Aufgabe (b): Steuerzeichen \b und \r testen ===

print("Status: FEHLERR\b ")
print("Ladevorgang...\rFERTIG")

# Was bewirkt \b?
# \b (Backspace) setzt den Cursor eine Position zurueck.
# Das naechste Zeichen (hier ein Leerzeichen) ueberschreibt
# das vorherige Zeichen. So wird das doppelte R korrigiert:
# "FEHLERR" wird zu "FEHLER ".

# Was bewirkt \r?
# \r (Carriage Return) setzt den Cursor an den Anfang der Zeile zurueck.
# Alles, was danach kommt, ueberschreibt den bisherigen Text von links.
# "Ladevorgang..." wird teilweise durch "FERTIG" ueberschrieben,
# Ausgabe: "FERTIGorgang..."
````

````python
# === Aufgabe (c): Ladebalken fuer den Systemstart ===
# import time und die while-Schleife sind vorgegeben.
# Die Loesung: \r am Anfang des Strings und end="" im print().

import time

print("Systemstart wird eingeleitet...")

fortschritt = 0
while fortschritt <= 100:
    balken = "#" * (fortschritt // 5)
    luecke = "." * (20 - fortschritt // 5)

    # Mit \r springt der Cursor an den Zeilenanfang zurueck.
    # Mit end="" wird kein Zeilenumbruch erzeugt.
    # So ueberschreibt jede Ausgabe die vorherige auf derselben Zeile.
    print("\r[" + balken + luecke + "] " + str(fortschritt) + "%", end="")

    time.sleep(0.1)
    fortschritt = fortschritt + 5

print("\nAlle Systeme bereit.")
````
