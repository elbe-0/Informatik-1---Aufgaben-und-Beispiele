## Briefing

Die Fabrikhalle der MotoTec GmbH liegt im Dunkeln. Seit dem IT-Totalausfall vor drei Stunden reagiert kein einziges System mehr - keine Beleuchtung, keine Maschinensteuerung, keine Kommunikation. Das IT-Notfallteam wurde gerufen und steht jetzt vor dem Hauptterminal im Serverraum. Sandra Holz, die IT-Leiterin, leuchtet mit einer Taschenlampe auf den Bildschirm. Er ist schwarz. Alles muss von Grund auf neu gestartet werden.

## Stakeholder-Dialog

*Sandra Holz (IT-Leiterin) steht angespannt am Terminal. Sie spricht knapp und ohne Umschweife:*

> *"Keine Zeit fuer Erklaerungen. Terminal an. Ich brauche eine Bootsequenz, die das System hochfaehrt und den Techniker registriert. Drei Dinge: Bootmeldung ausgeben, Name eingeben, Bestaetigung anzeigen. Los."*

## Aufgabe

Schreiben Sie ein Python-Programm, das die Notfall-Bootsequenz des MotoTec-Terminals durchfuehrt:

**(a)** Geben Sie eine Bootsequenz-Meldung auf dem Bildschirm aus. Das Terminal soll beim Hochfahren folgende Zeilen anzeigen:

```
=== MOTOTEC SYSTEMS v3.7 ===
Notfall-Boot wird eingeleitet...
```

Verwenden Sie dazu den `print()`-Befehl.

**(b)** Fragen Sie den Namen des Technikers ab. Das Terminal muss wissen, wer sich anmeldet. Verwenden Sie den `input()`-Befehl, um den Namen einzulesen und in einer Variable zu speichern.

**(c)** Geben Sie eine personalisierte Statusmeldung aus. Nutzen Sie den eingegebenen Namen, um eine Bestaetigung anzuzeigen, z.B.:

```
Techniker Max Mustermann angemeldet.
System bereit. Warte auf Anweisungen...
```

Verwenden Sie dazu String-Verkettung mit `+`.

---

**Hinweise:**
- `print("Text")` gibt Text auf dem Bildschirm aus
- `input("Frage: ")` wartet auf eine Eingabe und gibt den eingegebenen Text zurueck
- Mit `+` koennen Sie Strings (Texte) aneinanderhaengen: `"Hallo " + name`
- `print()` ohne Inhalt erzeugt eine Leerzeile

````python
# === Aufgabe (a): Bootsequenz ausgeben ===
# Mit print() geben wir Text auf dem Bildschirm aus.
# Jeder print()-Aufruf erzeugt eine neue Zeile.
print("=== MOTOTEC SYSTEMS v3.7 ===")
print("Notfall-Boot wird eingeleitet...")

# Eine Leerzeile fuer bessere Lesbarkeit
print()

# === Aufgabe (b): Techniker-Name eingeben ===
# input() zeigt einen Text an und wartet auf die Eingabe des Benutzers.
# Der eingegebene Text wird in der Variable 'name' gespeichert.
name = input("Techniker-Identifikation eingeben: ")

# === Aufgabe (c): Personalisierte Statusmeldung ausgeben ===
# Mit + verketten wir mehrere Strings (Texte) zu einem.
# Achtung: Leerzeichen muessen manuell eingefuegt werden!
print("Techniker " + name + " angemeldet.")
print("System bereit. Warte auf Anweisungen...")
````
