## Beispiele

Funktionen werden mit `def` definiert. Beim Aufruf müssen Anzahl und Reihenfolge der Argumente stimmen:

```python
# Funktion mit zwei Parametern
def maschine_starten(name, drehzahl):
    print(f"Maschine {name} startet mit {drehzahl} U/min.")

# Korrekter Aufruf: 2 Argumente für 2 Parameter
maschine_starten("Presswerk", 3000)
```

Was passiert bei falscher Anzahl von Argumenten?

```python
# TypeError: zu wenige Argumente
def motor_einstellen(typ, leistung, drehmoment):
    print(f"{typ}: {leistung} kW, {drehmoment} Nm")

# Aufruf mit nur 2 statt 3 Argumenten:
# motor_einstellen("Diesel", 150)  # -> TypeError!
```

Was passiert, wenn eine Funktion unter falschem Namen aufgerufen wird?

```python
# NameError: Funktion existiert nicht unter diesem Namen
def sensor_auslesen():
    print("Sensorwert: 42")

# Tippfehler im Aufruf:
# sensr_auslesen()  # -> NameError!
```

## Aufgabe

Der Roboterarm wird jetzt über Funktionen gesteuert. Analysieren Sie den Code und finden Sie die Fehler.

### (a) TypeError finden

Die Funktion `roboterarm_bewegen` erwartet drei Parameter. Der Aufruf unten ist fehlerhaft. Sagen Sie **vor dem Ausführen** vorher, welcher Fehler auftritt und warum.

### (b) Aufruf korrigieren

Korrigieren Sie den Funktionsaufruf, sodass er fehlerfrei läuft.

### (c) NameError finden

Im zweiten Code-Block wird eine Funktion aufgerufen, die nicht existiert. Finden Sie den Tippfehler und korrigieren Sie ihn.

```python
# === Aufgabe (a) und (b): TypeError analysieren und korrigieren ===
# Die Funktion erwartet 3 Parameter: richtung, geschwindigkeit, wiederholungen
# Der Aufruf übergibt nur 2 Argumente -> TypeError
# "missing 1 required positional argument: 'wiederholungen'"

def roboterarm_bewegen(richtung, geschwindigkeit, wiederholungen):
    print(f"Arm bewegt sich {wiederholungen}x nach {richtung} mit Tempo {geschwindigkeit}.")

# Fehlerhafter Aufruf (auskommentiert, da er einen Fehler erzeugt):
# roboterarm_bewegen("links", 50)  # TypeError!

# === Aufgabe (b): Aufruf korrigieren ===
# Alle drei Argumente übergeben:
roboterarm_bewegen("links", 50, 3)
```

```python
# === Aufgabe (c): NameError analysieren und korrigieren ===
# Die Funktion heißt greifer_öffnen() (mit ö),
# aber der Aufruf verwendet greifer_oeffnen() (mit oe).
# Python kennt keinen Zusammenhang zwischen ö und oe -> NameError

def greifer_öffnen():
    print("Greifer geöffnet.")

# Korrigierter Aufruf - gleicher Name wie die Definition:
greifer_öffnen()
```
