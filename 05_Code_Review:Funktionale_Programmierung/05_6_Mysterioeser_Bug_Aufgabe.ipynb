## Beispiele

### Lokaler vs. globaler Scope

Variablen, die innerhalb einer Funktion erstellt werden, existieren nur dort.
Sie sind **lokal**. Variablen außerhalb einer Funktion sind **global**:

````python
# Globale Variable
maschinentyp = "Fräse"

def zeige_status():
    # Lokale Variable - existiert nur in dieser Funktion
    status = "aktiv"
    print(f"Maschine: {maschinentyp}, Status: {status}")

zeige_status()
print(f"Maschinentyp: {maschinentyp}")
# print(status)  # Fehler! 'status' existiert hier nicht
````

### Das Schlüsselwort `global`

Mit `global` kann eine Funktion eine globale Variable direkt verändern.
Das ist gefährlich, weil es schwer nachvollziehbar macht, wo eine Variable
ihren Wert ändert:

````python
zaehler = 0

def erhoehe_zaehler():
    global zaehler  # Zugriff auf die globale Variable
    zaehler = zaehler + 1

print(f"Vorher: {zaehler}")
erhoehe_zaehler()
print(f"Nachher: {zaehler}")  # zaehler wurde verändert!
````

### Parameterweitergabe: Funktionen erhalten Kopien

Wenn Sie eine Variable an eine Funktion übergeben, arbeitet die Funktion
mit einer **Kopie** des Wertes. Änderungen am Parameter haben keinen
Einfluss auf die ursprüngliche Variable:

````python
def verdopple(wert):
    wert = wert * 2
    print(f"In der Funktion: {wert}")

temperatur = 80
verdopple(temperatur)
print(f"Nach dem Aufruf: {temperatur}")  # Unverändert!
````

## Hilfestellung

<details>
<summary>Hinweis zu (b): Warum sind die Werte unverändert?</summary>

Denken Sie daran: Wenn Sie eine Variable an eine Funktion übergeben,
erhält die Funktion eine Kopie des Wertes. Die Parameter `x` und `y`
in der Funktion sind eigene, lokale Variablen.

</details>

<details>
<summary>Hinweis zu (c): Lösung mit return</summary>

Eine Funktion kann mehrere Werte gleichzeitig zurückgeben:
```python
def funktion(a, b):
    return b, a
```
Beim Aufruf können Sie die Rückgabewerte direkt zuweisen:
```python
x, y = funktion(x, y)
```

<details>
<summary>Hinweis zu (d): Warum ist global gefährlich?</summary>

Mit `global` kann eine Funktion Variablen außerhalb ihres eigenen Bereichs
verändern. Das macht den Code unvorhersehbar: Jede Funktion könnte
beliebige globale Variablen ändern, ohne dass man es am Aufruf erkennt.
