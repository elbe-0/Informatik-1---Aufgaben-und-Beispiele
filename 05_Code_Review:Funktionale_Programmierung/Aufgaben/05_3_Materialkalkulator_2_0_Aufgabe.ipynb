## Beispiele

Wenn derselbe Code mehrfach vorkommt, sollte er in eine Funktion ausgelagert werden:

```python
# Vorher: Gleiche Berechnung an mehreren Stellen
gewicht_1 = 50
volumen_1 = gewicht_1 / 7.85  # Stahl-Dichte
print(f"Volumen: {volumen_1:.2f} l")

gewicht_2 = 120
volumen_2 = gewicht_2 / 7.85  # Gleiche Formel, wieder kopiert!
print(f"Volumen: {volumen_2:.2f} l")
```

```python
# Nachher: Funktion mit return
def berechne_volumen(gewicht, dichte):
    volumen = gewicht / dichte
    return volumen  # Ergebnis wird ZURÜCKGEGEBEN, nicht gedruckt

# Jetzt kann die Funktion beliebig oft aufgerufen werden:
v1 = berechne_volumen(50, 7.85)
v2 = berechne_volumen(120, 7.85)
print(f"Volumen 1: {v1:.2f} l")
print(f"Volumen 2: {v2:.2f} l")
```

Der Unterschied zwischen `print()` in der Funktion und `return`:
- `print()` gibt Text auf dem Bildschirm aus, aber die Funktion liefert `None` zurück
- `return` gibt einen Wert an den Aufrufer zurück, der weiterverarbeitet werden kann

```python
# print() vs return
def mit_print(a, b):
    print(a + b)  # Gibt aus, liefert aber None

def mit_return(a, b):
    return a + b  # Liefert das Ergebnis zurück

erg1 = mit_print(3, 4)    # Gibt "7" aus
print("Rückgabe:", erg1)   # None!

erg2 = mit_return(3, 4)   # Gibt nichts aus
print("Rückgabe:", erg2)   # 7 - kann weiterverarbeitet werden
```
