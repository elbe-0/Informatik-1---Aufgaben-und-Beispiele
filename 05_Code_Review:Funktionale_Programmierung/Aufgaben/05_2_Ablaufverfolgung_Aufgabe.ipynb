## Beispiele

Bei sequentiellem Code wird Zeile für Zeile von oben nach unten ausgeführt:

```python
# Sequentieller Ablauf: Zeile 1, 2, 3, 4
breite = 10
hoehe = 5
flaeche = breite * hoehe
print(f"Fläche: {flaeche} m²")
```

Bei Funktionsaufrufen springt die Ausführung in die Funktion und danach zurück:

```python
# Funktionsbasierter Ablauf:
# Zeile 1-3: Definition (wird NICHT sofort ausgeführt)
# Zeile 5: Aufruf -> springt in die Funktion (Zeile 2-3)
# Zeile 6: Ergebnis ausgeben

def berechne_flaeche(b, h):
    ergebnis = b * h
    return ergebnis

resultat = berechne_flaeche(10, 5)
print(f"Fläche: {resultat} m²")
```

Mit `print()`-Anweisungen können Sie die tatsächliche Ausführungsreihenfolge sichtbar machen:

```python
def berechne_flaeche(b, h):
    print(">>> Funktion berechne_flaeche gestartet")
    ergebnis = b * h
    print(">>> Funktion berechne_flaeche beendet")
    return ergebnis

print("Programm startet")
resultat = berechne_flaeche(10, 5)
print(f"Ergebnis: {resultat}")
print("Programm beendet")
```
