## Beispiele

### `datetime` importieren und aktuelle Zeit abrufen

Das Modul `datetime` stellt Funktionen für Datum und Uhrzeit bereit.
Mit `datetime.datetime.now()` erhalten Sie den aktuellen Zeitpunkt:

```python
import datetime

# Aktuelle Zeit abrufen
jetzt = datetime.datetime.now()
print("Aktueller Zeitpunkt:", jetzt)
```

### Zeit formatieren mit `strftime()`

Die Methode `strftime()` formatiert einen Zeitpunkt als lesbaren Text.
Wichtige Platzhalter:
- `%Y` - Jahr (vierstellig)
- `%m` - Monat (zweistellig)
- `%d` - Tag (zweistellig)
- `%H` - Stunde (24h-Format)
- `%M` - Minute
- `%S` - Sekunde

```python
import datetime

jetzt = datetime.datetime.now()

# Verschiedene Formate für das Maschinenprotokoll
print(jetzt.strftime("%Y-%m-%d"))           # z.B. 2024-01-15
print(jetzt.strftime("%H:%M:%S"))           # z.B. 14:30:00
print(jetzt.strftime("%Y-%m-%d %H:%M:%S"))  # z.B. 2024-01-15 14:30:00
```

### Funktion mit `return`

Eine Funktion kann mit `return` einen Wert zurückgeben. Der Aufrufer
kann diesen Wert in einer Variablen speichern und weiterverwenden:

```python
def berechne_durchsatz(stueckzahl, stunden):
    """Berechnet den Durchsatz pro Stunde und gibt ihn zurück."""
    durchsatz = stueckzahl / stunden
    return durchsatz

# Rückgabewert in Variable speichern
ergebnis = berechne_durchsatz(480, 8)
print(f"Durchsatz: {ergebnis} Stück/Stunde")
```
