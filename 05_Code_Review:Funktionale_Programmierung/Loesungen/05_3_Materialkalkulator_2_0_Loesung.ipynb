## Beispiele

Wenn derselbe Code mehrfach vorkommt, sollte er in eine Funktion ausgelagert werden:

```python
# Vorher: Gleiche Berechnung an mehreren Stellen
gewicht_1 = 50
volumen_1 = gewicht_1 / 7.85  # Stahl-Dichte
print(f"Volumen: {volumen_1:.2f} l")

gewicht_2 = 120
volumen_2 = gewicht_2 / 7.85  # Gleiche Formel, wieder kopiert!
print(f"Volumen: {volumen_2:.2f} l")
```

```python
# Nachher: Funktion mit return
def berechne_volumen(gewicht, dichte):
    volumen = gewicht / dichte
    return volumen  # Ergebnis wird ZURÜCKGEGEBEN, nicht gedruckt

# Jetzt kann die Funktion beliebig oft aufgerufen werden:
v1 = berechne_volumen(50, 7.85)
v2 = berechne_volumen(120, 7.85)
print(f"Volumen 1: {v1:.2f} l")
print(f"Volumen 2: {v2:.2f} l")
```

Der Unterschied zwischen `print()` in der Funktion und `return`:
- `print()` gibt Text auf dem Bildschirm aus, aber die Funktion liefert `None` zurück
- `return` gibt einen Wert an den Aufrufer zurück, der weiterverarbeitet werden kann

```python
# print() vs return
def mit_print(a, b):
    print(a + b)  # Gibt aus, liefert aber None

def mit_return(a, b):
    return a + b  # Liefert das Ergebnis zurück

erg1 = mit_print(3, 4)    # Gibt "7" aus
print("Rückgabe:", erg1)   # None!

erg2 = mit_return(3, 4)   # Gibt nichts aus
print("Rückgabe:", erg2)   # 7 - kann weiterverarbeitet werden
```

## Aufgabe

Refactoren Sie den Materialkalkulator aus Kapitel 3 in eine wiederverwendbare Funktion.

### (a) Alten Code analysieren

Der folgende Code ist der Materialkalkulator aus Aufgabe 3.6. Identifizieren Sie: Welcher Teil könnte eine Funktion werden?

### (b) Funktion schreiben

Schreiben Sie eine Funktion `berechne_material(menge_a, menge_b, operation)`, die zwei Mengen und einen Operator entgegennimmt, die Berechnung durchführt und das Ergebnis mit `return` **zurückgibt** (nicht mit `print()` ausgibt).

### (c) Funktion mit Eingabe verwenden

Lesen Sie zwei Mengen und eine Operation per `input()` ein und rufen Sie die Funktion auf. Geben Sie das Ergebnis mit `print()` aus.

### (d) Wiederverwendbarkeit demonstrieren

Rufen Sie die Funktion dreimal mit verschiedenen festen Werten auf (ohne `input()`), um zu zeigen, dass sie wiederverwendbar ist.

```python
# === Alter Materialkalkulator aus Kapitel 3 (nicht verändern) ===

menge_a = float(input("Erste Materialmenge: "))
menge_b = float(input("Zweite Materialmenge: "))
operation = input("Rechenoperation (+, -, *, /): ")

if operation == "+":
    print("Ergebnis:", menge_a + menge_b)
elif operation == "-":
    print("Ergebnis:", menge_a - menge_b)
elif operation == "*":
    print("Ergebnis:", menge_a * menge_b)
elif operation == "/":
    print("Ergebnis:", menge_a / menge_b)
else:
    print("Ungültige Operation!")
```

```python
# === Aufgabe (a): Analyse des alten Codes ===
# Der gesamte if/elif/else-Block ist die eigentliche Berechnung.
# Die Eingabe (input) und die Ausgabe (print) sind getrennte Aufgaben.
# -> Die Berechnung wird zur Funktion, Ein- und Ausgabe bleiben außerhalb.
```

```python
# === Aufgabe (b): Funktion mit return ===
# Die Funktion nimmt die Werte entgegen, berechnet das Ergebnis
# und gibt es mit return zurück. Kein print() in der Funktion!

def berechne_material(menge_a, menge_b, operation):
    if operation == "+":
        return menge_a + menge_b
    elif operation == "-":
        return menge_a - menge_b
    elif operation == "*":
        return menge_a * menge_b
    elif operation == "/":
        return menge_a / menge_b
    else:
        return None  # Ungültige Operation
```

```python
# === Aufgabe (c): Funktion mit Eingabe verwenden ===
menge_a = float(input("Erste Materialmenge: "))
menge_b = float(input("Zweite Materialmenge: "))
operation = input("Rechenoperation (+, -, *, /): ")

ergebnis = berechne_material(menge_a, menge_b, operation)
if ergebnis is not None:
    print(f"Ergebnis: {ergebnis}")
else:
    print("Ungültige Operation!")
```

```python
# === Aufgabe (d): Wiederverwendbarkeit demonstrieren ===
# Drei Aufrufe mit verschiedenen Werten - ohne input()!
print("750.5 + 320.0 =", berechne_material(750.5, 320.0, "+"))
print("1000 - 247.3 =", berechne_material(1000, 247.3, "-"))
print("12.5 * 8 =", berechne_material(12.5, 8, "*"))
```
