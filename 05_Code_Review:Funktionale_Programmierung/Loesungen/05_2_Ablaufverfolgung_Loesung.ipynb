## Beispiele

Bei sequentiellem Code wird Zeile für Zeile von oben nach unten ausgeführt:

```python
# Sequentieller Ablauf: Zeile 1, 2, 3, 4
breite = 10
hoehe = 5
flaeche = breite * hoehe
print(f"Fläche: {flaeche} m²")
```

Bei Funktionsaufrufen springt die Ausführung in die Funktion und danach zurück:

```python
# Funktionsbasierter Ablauf:
# Zeile 1-3: Definition (wird NICHT sofort ausgeführt)
# Zeile 5: Aufruf -> springt in die Funktion (Zeile 2-3)
# Zeile 6: Ergebnis ausgeben

def berechne_flaeche(b, h):
    ergebnis = b * h
    return ergebnis

resultat = berechne_flaeche(10, 5)
print(f"Fläche: {resultat} m²")
```

Mit `print()`-Anweisungen können Sie die tatsächliche Ausführungsreihenfolge sichtbar machen:

```python
def berechne_flaeche(b, h):
    print(">>> Funktion berechne_flaeche gestartet")
    ergebnis = b * h
    print(">>> Funktion berechne_flaeche beendet")
    return ergebnis

print("Programm startet")
resultat = berechne_flaeche(10, 5)
print(f"Ergebnis: {resultat}")
print("Programm beendet")
```

## Aufgabe

Die Maschinensteuerung berechnet die tägliche Laufzeit einer Maschine. Analysieren Sie beide Varianten.

### (a) Sequentieller Code

Lesen Sie den folgenden Code. Notieren Sie **vor dem Ausführen**: Was wird ausgegeben? In welcher Reihenfolge werden die Zeilen ausgeführt?

### (b) Funktionsbasierter Code

Die gleiche Berechnung wurde in Funktionen aufgeteilt. In welcher Reihenfolge werden die Zeilen jetzt ausgeführt? Wann springt Python in eine Funktion, wann zurück?

### (c) Trace-Ausgaben

Fügen Sie `print()`-Anweisungen am Anfang und Ende jeder Funktion ein, um die Ausführungsreihenfolge sichtbar zu machen.

```python
# === Aufgabe (a): Sequentieller Code ===
# Ausführungsreihenfolge: Zeile für Zeile von oben nach unten
# Zeile 1: start_stunde = 6
# Zeile 2: ende_stunde = 22
# Zeile 3: laufzeit = 22 - 6 = 16
# Zeile 4: print -> "Maschine lief 16 Stunden."
# Zeile 5: print -> "Das sind 960 Minuten."

start_stunde = 6
ende_stunde = 22
laufzeit = ende_stunde - start_stunde
print(f"Maschine lief {laufzeit} Stunden.")
print(f"Das sind {laufzeit * 60} Minuten.")
```

```python
# === Aufgabe (b): Funktionsbasierter Code ===
# Ausführungsreihenfolge:
# 1. def berechne_laufzeit -> wird nur DEFINIERT, nicht ausgeführt
# 2. def laufzeit_abfragen -> wird nur DEFINIERT, nicht ausgeführt
# 3. laufzeit_abfragen() wird AUFGERUFEN -> springt in die Funktion
#    3a. input() für Startzeit
#    3b. input() für Endzeit
#    3c. return -> springt zurück
# 4. berechne_laufzeit(start, ende) wird AUFGERUFEN
#    4a. Berechnung -> return -> springt zurück
# 5. print() gibt das Ergebnis aus

def berechne_laufzeit(start, ende):
    return ende - start

def laufzeit_abfragen():
    s = int(input("Startzeit (Stunde): "))
    e = int(input("Endzeit (Stunde): "))
    return s, e

start, ende = laufzeit_abfragen()
stunden = berechne_laufzeit(start, ende)
print(f"Maschine lief {stunden} Stunden.")
```

```python
# === Aufgabe (c): Trace-Ausgaben ===
# print()-Anweisungen am Anfang und Ende jeder Funktion
# machen die Ausführungsreihenfolge sichtbar.

def berechne_laufzeit(start, ende):
    print(">>> berechne_laufzeit gestartet")
    ergebnis = ende - start
    print(">>> berechne_laufzeit beendet")
    return ergebnis

def laufzeit_abfragen():
    print(">>> laufzeit_abfragen gestartet")
    s = int(input("Startzeit (Stunde): "))
    e = int(input("Endzeit (Stunde): "))
    print(">>> laufzeit_abfragen beendet")
    return s, e

print("Programm startet")
start, ende = laufzeit_abfragen()
stunden = berechne_laufzeit(start, ende)
print(f"Maschine lief {stunden} Stunden.")
print("Programm beendet")
```
