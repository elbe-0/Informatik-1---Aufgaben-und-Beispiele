## Beispiele

### `datetime` importieren und aktuelle Zeit abrufen

Das Modul `datetime` stellt Funktionen für Datum und Uhrzeit bereit.
Mit `datetime.datetime.now()` erhalten Sie den aktuellen Zeitpunkt:

```python
import datetime

# Aktuelle Zeit abrufen
jetzt = datetime.datetime.now()
print("Aktueller Zeitpunkt:", jetzt)
```

### Zeit formatieren mit `strftime()`

Die Methode `strftime()` formatiert einen Zeitpunkt als lesbaren Text.
Wichtige Platzhalter:
- `%Y` - Jahr (vierstellig)
- `%m` - Monat (zweistellig)
- `%d` - Tag (zweistellig)
- `%H` - Stunde (24h-Format)
- `%M` - Minute
- `%S` - Sekunde

```python
import datetime

jetzt = datetime.datetime.now()

# Verschiedene Formate für das Maschinenprotokoll
print(jetzt.strftime("%Y-%m-%d"))           # z.B. 2024-01-15
print(jetzt.strftime("%H:%M:%S"))           # z.B. 14:30:00
print(jetzt.strftime("%Y-%m-%d %H:%M:%S"))  # z.B. 2024-01-15 14:30:00
```

### Funktion mit `return`

Eine Funktion kann mit `return` einen Wert zurückgeben. Der Aufrufer
kann diesen Wert in einer Variablen speichern und weiterverwenden:

```python
def berechne_durchsatz(stueckzahl, stunden):
    """Berechnet den Durchsatz pro Stunde und gibt ihn zurück."""
    durchsatz = stueckzahl / stunden
    return durchsatz

# Rückgabewert in Variable speichern
ergebnis = berechne_durchsatz(480, 8)
print(f"Durchsatz: {ergebnis} Stück/Stunde")
```

## Aufgabe

Erstellen Sie Funktionen für das Schichtprotokoll der Fabrik:

**(a)** Schreiben Sie eine Funktion `get_aktuelle_zeit()`, die die aktuelle
Uhrzeit mit `datetime.datetime.now()` ermittelt und als Rückgabewert zurückgibt.

**(b)** Schreiben Sie eine Funktion `zeit_ausgabe(zeit)`, die einen Zeitpunkt entgegennimmt
und ihn im Format `"[2024-01-15 14:30:00] Schichtprotokoll-Eintrag"` ausgibt.
Verwenden Sie `strftime()` für die Formatierung.

**(c)** Nutzen Sie beide Funktionen zusammen: Rufen Sie `get_aktuelle_zeit()` auf,
speichern Sie das Ergebnis, und übergeben Sie es an `zeit_ausgabe()`.

**(d)** Schreiben Sie eine Funktion `protokoll_eintrag(aktion)`, die beide Schritte
kombiniert: Sie ermittelt die aktuelle Zeit und gibt sie zusammen mit der übergebenen
Aktion aus, z.B. `"[2024-01-15 14:30:00] Maschine 3 gestartet"`. Testen Sie mit
`protokoll_eintrag("Maschine 3 gestartet")`.

```python
import datetime

# === Aufgabe (a): Funktion get_aktuelle_zeit() ===

def get_aktuelle_zeit():
    """Gibt die aktuelle Uhrzeit zurück."""
    return datetime.datetime.now()


# === Aufgabe (b): Funktion zeit_ausgabe(zeit) ===

def zeit_ausgabe(zeit):
    """Gibt die Zeit formatiert als Schichtprotokoll-Eintrag aus."""
    formatiert = zeit.strftime("%Y-%m-%d %H:%M:%S")
    print(f"[{formatiert}] Schichtprotokoll-Eintrag")


# === Aufgabe (c): Beide Funktionen zusammen nutzen ===

zeit = get_aktuelle_zeit()
zeit_ausgabe(zeit)


# === Aufgabe (d): Funktion protokoll_eintrag(aktion) ===

def protokoll_eintrag(aktion):
    """Ermittelt die aktuelle Zeit und gibt sie mit der Aktion aus."""
    zeit = get_aktuelle_zeit()
    formatiert = zeit.strftime("%Y-%m-%d %H:%M:%S")
    print(f"[{formatiert}] {aktion}")

protokoll_eintrag("Maschine 3 gestartet")
```
