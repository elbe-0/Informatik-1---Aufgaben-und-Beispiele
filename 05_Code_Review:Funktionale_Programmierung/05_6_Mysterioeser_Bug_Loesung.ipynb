## Beispiele

### Lokaler vs. globaler Scope

Variablen, die innerhalb einer Funktion erstellt werden, existieren nur dort.
Sie sind **lokal**. Variablen außerhalb einer Funktion sind **global**:

```python
# Globale Variable
maschinentyp = "Fräse"

def zeige_status():
    # Lokale Variable - existiert nur in dieser Funktion
    status = "aktiv"
    print(f"Maschine: {maschinentyp}, Status: {status}")

zeige_status()
print(f"Maschinentyp: {maschinentyp}")
# print(status)  # Fehler! 'status' existiert hier nicht
```

### Das Schlüsselwort `global`

Mit `global` kann eine Funktion eine globale Variable direkt verändern.
Das ist gefährlich, weil es schwer nachvollziehbar macht, wo eine Variable
ihren Wert ändert:

```python
zaehler = 0

def erhoehe_zaehler():
    global zaehler  # Zugriff auf die globale Variable
    zaehler = zaehler + 1

print(f"Vorher: {zaehler}")
erhoehe_zaehler()
print(f"Nachher: {zaehler}")  # zaehler wurde verändert!
```

### Parameterweitergabe: Funktionen erhalten Kopien

Wenn Sie eine Variable an eine Funktion übergeben, arbeitet die Funktion
mit einer **Kopie** des Wertes. Änderungen am Parameter haben keinen
Einfluss auf die ursprüngliche Variable:

```python
def verdopple(wert):
    wert = wert * 2
    print(f"In der Funktion: {wert}")

temperatur = 80
verdopple(temperatur)
print(f"Nach dem Aufruf: {temperatur}")  # Unverändert!
```

## Aufgabe

Ein Kollege hat folgende Funktion geschrieben, um die Soll-Temperatur und die
Ist-Temperatur zu tauschen. Aber es funktioniert nicht wie erwartet.

**(a)** Analysieren Sie den folgenden Code. Sagen Sie **vorher**, was ausgegeben wird,
**bevor** Sie die Zelle ausführen.

**(b)** Warum sind `soll` und `ist` nach dem Funktionsaufruf unverändert?
Erklären Sie den Grund in einem Kommentar.

**(c)** Beheben Sie das Problem: Schreiben Sie eine Version von `tausche_werte(x, y)`,
die mit `return` die getauschten Werte zurückgibt, und rufen Sie die Funktion so auf,
dass die Werte tatsächlich getauscht werden.

**(d)** Gefahrenzone: Ein Kollege schlägt vor, `global` zu verwenden. Zeigen Sie,
was passiert, wenn Sie `global soll` in der Funktion nutzen. Erklären Sie in
einem Kommentar, warum das gefährlich ist.

```python
# === Aufgabe (a): Code analysieren und Vorhersage ===

def tausche_werte(x, y):
    temp = x
    x = y
    y = temp
    print(f"In der Funktion: x={x}, y={y}")

soll = 80
ist = 95
tausche_werte(soll, ist)
print(f"Nach dem Aufruf: soll={soll}, ist={ist}")

# Vorhersage:
# In der Funktion: x=95, y=80
# Nach dem Aufruf: soll=80, ist=95
# -> Die Werte werden nur innerhalb der Funktion getauscht,
#    außerhalb bleiben sie unverändert.
```

```python
# === Aufgabe (b): Erklärung ===

# Die Funktion erhält Kopien der Werte von soll und ist.
# Die Parameter x und y sind lokale Variablen innerhalb der Funktion.
# Wenn x und y getauscht werden, betrifft das nur diese lokalen Kopien.
# Die originalen Variablen soll und ist außerhalb der Funktion
# werden davon nicht berührt - sie behalten ihre ursprünglichen Werte.
```

```python
# === Aufgabe (c): Version mit return ===

def tausche_werte(x, y):
    """Tauscht zwei Werte und gibt sie zurück."""
    return y, x

soll = 80
ist = 95
print(f"Vorher: soll={soll}, ist={ist}")

soll, ist = tausche_werte(soll, ist)
print(f"Nachher: soll={soll}, ist={ist}")
```

```python
# === Aufgabe (d): Gefahrenzone - Version mit global ===

soll = 80
ist = 95

def tausche_werte_global():
    global soll
    global ist
    temp = soll
    soll = ist
    ist = temp

print(f"Vorher: soll={soll}, ist={ist}")
tausche_werte_global()
print(f"Nachher: soll={soll}, ist={ist}")

# WARNUNG: global ist gefährlich, weil:
# - Die Funktion verändert Variablen außerhalb ihres eigenen Bereichs.
# - Am Funktionsaufruf ist nicht erkennbar, welche Variablen verändert werden.
# - Bei großen Programmen wird der Code unvorhersehbar: Jede Funktion könnte
#   beliebige globale Variablen ändern, ohne dass man es bemerkt.
# - Besser: Werte als Parameter übergeben und mit return zurückgeben.
```
