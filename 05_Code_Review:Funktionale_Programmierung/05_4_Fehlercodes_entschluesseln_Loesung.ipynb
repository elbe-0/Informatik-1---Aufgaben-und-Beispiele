## Beispiele

Mit Ganzzahldivision `//` und Modulo `%` kann man Ziffern einer Zahl extrahieren:

```python
# Ganzzahldivision: schneidet den Rest ab
print(13 // 2)  # 6 (nicht 6.5)
print(13 % 2)   # 1 (der Rest)

# Wiederholtes Teilen durch 2 ergibt die Binärziffern:
zahl = 13
print(zahl % 2)   # 1 (letzte Binärziffer)
zahl = zahl // 2  # 6
print(zahl % 2)   # 0
zahl = zahl // 2  # 3
print(zahl % 2)   # 1
zahl = zahl // 2  # 1
print(zahl % 2)   # 1
# Von unten nach oben gelesen: 1101 (= 13 in binär)
```

Mit `list.insert(0, wert)` können Sie Werte am Anfang einer Liste einfügen:

```python
# insert(0, x) fügt am Anfang ein (statt am Ende)
ziffern = []
ziffern.insert(0, 1)  # [1]
ziffern.insert(0, 0)  # [0, 1]
ziffern.insert(0, 1)  # [1, 0, 1]
print(ziffern)  # [1, 0, 1] - die Ziffern in richtiger Reihenfolge
```

## Aufgabe

Schreiben Sie eine Funktion, die Dezimalzahlen in Binärzahlen umwandelt.

### (a) Funktion schreiben

Schreiben Sie eine Funktion `dezimal_zu_binaer(dezimalzahl)`, die eine Dezimalzahl
in eine Binärzahl umwandelt und das Ergebnis als Ganzzahl zurückgibt.
Nutzen Sie den Algorithmus aus dem Beispiel oben.

### (b) Testen

Testen Sie die Funktion mit den Fehlercodes 42, 127 und 255.

### (c) Handrechnung

Überprüfen Sie das Ergebnis für 42 per Handrechnung: Teilen Sie wiederholt durch 2 und notieren Sie die Reste.

```python
# === Aufgabe (a): Funktion dezimal_zu_binaer ===
# Algorithmus: Wiederholtes Teilen durch 2.
# Der Rest (% 2) ergibt die Binärziffern von rechts nach links.
# Mit insert(0, ...) werden sie in der richtigen Reihenfolge aufgebaut.

def dezimal_zu_binaer(dezimalzahl):
    binaer_liste = []
    binaer_zahl = 0

    while dezimalzahl > 0:
        rest = dezimalzahl % 2
        binaer_liste.insert(0, rest)
        dezimalzahl = dezimalzahl // 2

    # Liste in eine zusammenhängende Zahl umwandeln
    for ziffer in binaer_liste:
        binaer_zahl = binaer_zahl * 10 + ziffer

    return binaer_zahl
```

```python
# === Aufgabe (b): Testen mit Fehlercodes ===
for code in [42, 127, 255]:
    binaer = dezimal_zu_binaer(code)
    print(f"Fehlercode {code} -> Binär: {binaer}")
```

```python
# === Aufgabe (c): Handrechnung für 42 ===
# 42 // 2 = 21, Rest 0
# 21 // 2 = 10, Rest 1
# 10 // 2 =  5, Rest 0
#  5 // 2 =  2, Rest 1
#  2 // 2 =  1, Rest 0
#  1 // 2 =  0, Rest 1
# Reste von unten nach oben: 101010
# -> 42 dezimal = 101010 binär ✓
print("Kontrolle: dezimal_zu_binaer(42) =", dezimal_zu_binaer(42))
```
