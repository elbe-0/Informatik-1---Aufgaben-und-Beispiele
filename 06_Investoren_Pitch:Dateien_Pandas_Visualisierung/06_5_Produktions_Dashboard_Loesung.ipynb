## Beispiele

Die Bibliothek `matplotlib` bietet verschiedene Diagrammtypen. Hier einige Grundlagen:

### Balkendiagramm mit `plt.bar()`

```python
import matplotlib.pyplot as plt

# Einfaches Balkendiagramm
abteilungen = ["Fertigung", "Logistik", "Qualität"]
mitarbeiter = [45, 22, 18]

plt.bar(abteilungen, mitarbeiter, color="steelblue")
plt.title("Mitarbeiter pro Abteilung")
plt.ylabel("Anzahl")
plt.show()
```

### Streudiagramm mit `plt.scatter()`

```python
# Streudiagramm mit Farbe und Legende
temperatur = [20, 22, 25, 28, 30, 18, 21, 26]
leistung = [80, 82, 88, 91, 85, 78, 83, 90]
schicht = ["Tag", "Tag", "Tag", "Tag", "Nacht", "Nacht", "Nacht", "Nacht"]

farben = ["orange" if s == "Tag" else "navy" for s in schicht]

plt.scatter(temperatur, leistung, c=farben)
plt.xlabel("Temperatur (°C)")
plt.ylabel("Leistung (%)")
plt.title("Temperatur vs. Leistung")
plt.show()
```

### Histogramm mit `plt.hist()`

```python
# Histogramm: Verteilung von Messwerten
messwerte = [12, 15, 14, 13, 16, 15, 14, 17, 13, 12, 15, 16, 14, 15, 13]

plt.hist(messwerte, bins=6, color="seagreen", edgecolor="black")
plt.title("Verteilung der Messwerte")
plt.xlabel("Messwert")
plt.ylabel("Häufigkeit")
plt.show()
```

### Mehrere Diagramme mit `plt.subplots()`

```python
# Zwei Diagramme nebeneinander in einem Figure
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

ax1.bar(["A", "B", "C"], [10, 20, 15], color="coral")
ax1.set_title("Diagramm 1")

ax2.hist([1, 2, 2, 3, 3, 3, 4, 4, 5], bins=5, color="skyblue", edgecolor="black")
ax2.set_title("Diagramm 2")

plt.tight_layout()
plt.show()
```

### Diagramm speichern mit `plt.savefig()`

```python
# Diagramm als PNG speichern
plt.bar(["X", "Y", "Z"], [5, 8, 3], color="mediumpurple")
plt.title("Beispiel-Export")
plt.savefig("beispiel.png", dpi=150, bbox_inches="tight")
plt.show()
print("Datei 'beispiel.png' wurde gespeichert.")
```

## Aufgabe

Erstellen Sie ein Produktions-Dashboard, das alle wichtigen Kennzahlen aus der Datei `autos.csv` auf einer Seite darstellt.

### (a) Balkendiagramm: Produktion pro Marke

Lesen Sie die Datei `autos.csv` mit Pandas ein und erstellen Sie ein Balkendiagramm, das die Anzahl der Fahrzeuge pro Marke zeigt.

### (b) Streudiagramm: Verbrauch vs. CO2-Ausstoß

Erstellen Sie ein Streudiagramm, das den Kraftstoffverbrauch (x-Achse) gegen den CO2-Ausstoß (y-Achse) zeigt. Jede Marke soll eine eigene Farbe haben. Fügen Sie eine Legende hinzu.

### (c) Histogramm: Verteilung der Baujahre

Erstellen Sie ein Histogramm, das die Verteilung der Baujahre zeigt.

### (d) Dashboard: Alle Diagramme in einem Figure

Kombinieren Sie die drei Diagramme aus (a), (b) und (c) als Subplots in einem einzigen Figure und speichern Sie das Ergebnis als `dashboard.png`.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Daten einlesen
df = pd.read_csv("autos.csv")

# === Aufgabe (a): Balkendiagramm – Produktion pro Marke ===
marken_anzahl = df["Marke"].value_counts()

plt.bar(marken_anzahl.index, marken_anzahl.values, color="steelblue")
plt.title("Produktion pro Marke")
plt.xlabel("Marke")
plt.ylabel("Anzahl Fahrzeuge")
plt.show()
```

```python
# === Aufgabe (b): Streudiagramm – Verbrauch vs. CO2-Ausstoß ===
farben_map = {
    "BMW": "blue",
    "Mercedes": "silver",
    "Audi": "red",
    "VW": "darkblue",
    "Toyota": "green"
}

for marke, farbe in farben_map.items():
    daten = df[df["Marke"] == marke]
    plt.scatter(
        daten["Kraftstoffverbrauch (l/100km)"],
        daten["CO2-Ausstoß (g/km)"],
        c=farbe,
        label=marke,
        alpha=0.6
    )

plt.title("Verbrauch vs. CO2-Ausstoß")
plt.xlabel("Kraftstoffverbrauch (l/100km)")
plt.ylabel("CO2-Ausstoß (g/km)")
plt.legend()
plt.show()
```

```python
# === Aufgabe (c): Histogramm – Verteilung der Baujahre ===
plt.hist(df["Baujahr"], bins=7, color="seagreen", edgecolor="black")
plt.title("Verteilung der Baujahre")
plt.xlabel("Baujahr")
plt.ylabel("Häufigkeit")
plt.show()
```

```python
# === Aufgabe (d): Dashboard – Subplots + Speichern ===
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 5))

# Subplot 1: Balkendiagramm
marken_anzahl = df["Marke"].value_counts()
ax1.bar(marken_anzahl.index, marken_anzahl.values, color="steelblue")
ax1.set_title("Produktion pro Marke")
ax1.set_xlabel("Marke")
ax1.set_ylabel("Anzahl Fahrzeuge")

# Subplot 2: Streudiagramm
farben_map = {
    "BMW": "blue",
    "Mercedes": "silver",
    "Audi": "red",
    "VW": "darkblue",
    "Toyota": "green"
}
for marke, farbe in farben_map.items():
    daten = df[df["Marke"] == marke]
    ax2.scatter(
        daten["Kraftstoffverbrauch (l/100km)"],
        daten["CO2-Ausstoß (g/km)"],
        c=farbe,
        label=marke,
        alpha=0.6
    )
ax2.set_title("Verbrauch vs. CO2-Ausstoß")
ax2.set_xlabel("Kraftstoffverbrauch (l/100km)")
ax2.set_ylabel("CO2-Ausstoß (g/km)")
ax2.legend()

# Subplot 3: Histogramm
ax3.hist(df["Baujahr"], bins=7, color="seagreen", edgecolor="black")
ax3.set_title("Verteilung der Baujahre")
ax3.set_xlabel("Baujahr")
ax3.set_ylabel("Häufigkeit")

plt.tight_layout()
plt.savefig("dashboard.png", dpi=150, bbox_inches="tight")
plt.show()
print("Dashboard wurde als 'dashboard.png' gespeichert.")
```
