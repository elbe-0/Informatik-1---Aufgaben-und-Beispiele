## Beispiele

Plotly Express erzeugt interaktive Diagramme aus DataFrames. Hier ein einfaches Liniendiagramm mit Beispieldaten:

```python
import plotly.express as px
import pandas as pd

# Beispieldaten als DataFrame
daten = pd.DataFrame({
    "Monat": ["Jan", "Feb", "Mär", "Apr", "Mai"],
    "Umsatz": [120, 135, 150, 142, 160]
})

# Liniendiagramm erstellen
fig = px.line(daten, x="Monat", y="Umsatz", title="Umsatzentwicklung")
fig.show()
```

Mit `scatter_3d()` lassen sich drei Dimensionen gleichzeitig darstellen. Der Parameter `color` färbt die Punkte nach einer Kategorie ein:

```python
# 3D-Scatterplot mit Beispieldaten
daten_3d = pd.DataFrame({
    "X": [1, 2, 3, 4, 5, 6],
    "Y": [10, 20, 15, 25, 30, 20],
    "Z": [5, 8, 6, 9, 7, 10],
    "Gruppe": ["A", "A", "B", "B", "A", "B"]
})

fig = px.scatter_3d(daten_3d, x="X", y="Y", z="Z", color="Gruppe",
                    title="3D-Scatterplot")
fig.show()
```

Einen DataFrame können Sie nach einer Spalte filtern und mit `sort_values()` sortieren:

```python
# CSV einlesen und nach Datum sortieren
df = pd.read_csv("temperatures.csv")
df_sortiert = df.sort_values(by="Datum")

# Nur eine bestimmte Anlage filtern
df_m1 = df_sortiert[df_sortiert["Anlage"] == "Machine 1"]
print(f"{len(df_m1)} Datenpunkte für Machine 1")
df_m1.head()
```
