## Beispiele

Plotly Express erzeugt interaktive Diagramme aus DataFrames. Hier ein einfaches Liniendiagramm mit Beispieldaten:

```python
import plotly.express as px
import pandas as pd

# Beispieldaten als DataFrame
daten = pd.DataFrame({
    "Monat": ["Jan", "Feb", "Mär", "Apr", "Mai"],
    "Umsatz": [120, 135, 150, 142, 160]
})

# Liniendiagramm erstellen
fig = px.line(daten, x="Monat", y="Umsatz", title="Umsatzentwicklung")
fig.show()
```

Mit `scatter_3d()` lassen sich drei Dimensionen gleichzeitig darstellen. Der Parameter `color` färbt die Punkte nach einer Kategorie ein:

```python
# 3D-Scatterplot mit Beispieldaten
daten_3d = pd.DataFrame({
    "X": [1, 2, 3, 4, 5, 6],
    "Y": [10, 20, 15, 25, 30, 20],
    "Z": [5, 8, 6, 9, 7, 10],
    "Gruppe": ["A", "A", "B", "B", "A", "B"]
})

fig = px.scatter_3d(daten_3d, x="X", y="Y", z="Z", color="Gruppe",
                    title="3D-Scatterplot")
fig.show()
```

## Aufgabe

Erstellen Sie ein interaktives Dashboard aus den Maschinentemperaturen.

### (a) Temperaturverlauf als Liniendiagramm

Lesen Sie `temperatures.csv` ein und sortieren Sie die Daten nach Datum. Filtern Sie die Daten auf eine einzelne Anlage (z.B. "Machine 1"). Erstellen Sie ein Liniendiagramm mit Plotly Express, das den Temperaturverlauf dieser Anlage über die Zeit zeigt.

### (b) 3D-Scatterplot

Erstellen Sie einen 3D-Scatterplot über den gesamten Datensatz (alle Anlagen). Verwenden Sie das Datum als X-Achse, die Temperatur als Y-Achse und die Luftfeuchtigkeit als Z-Achse. Färben Sie die Datenpunkte nach Anlage ein.

### (c) Dashboard verfeinern

Geben Sie beiden Diagrammen aussagekräftige Titel, die für Investoren geeignet sind.

```python
import pandas as pd
import plotly.express as px

# === Aufgabe (a): Temperaturverlauf als Liniendiagramm ===
df = pd.read_csv("temperatures.csv")
df_sortiert = df.sort_values(by="Datum")

# Auf Machine 1 filtern
df_m1 = df_sortiert[df_sortiert["Anlage"] == "Machine 1"]

# Liniendiagramm erstellen
fig1 = px.line(df_m1, x="Datum", y="Temperatur",
               title="Temperaturverlauf Machine 1 - Produktionsüberwachung")
fig1.show()
```

```python
# === Aufgabe (b): 3D-Scatterplot ===
fig2 = px.scatter_3d(df_sortiert, x="Datum", y="Temperatur", z="Luftfeuchtigkeit",
                     color="Anlage",
                     title="Temperatur und Luftfeuchtigkeit aller Anlagen")
fig2.show()
```

```python
# === Aufgabe (c): Dashboard verfeinern ===
# Die Titel wurden bereits in (a) und (b) investorengerecht gewählt.
# Alternativ können die Titel nachträglich geändert werden:
fig1.update_layout(title="MotoTec Machine 1 - Temperaturstabilität im Produktionsbetrieb")
fig1.show()

fig2.update_layout(title="MotoTec Maschinenpark - Klima-Monitoring aller Anlagen")
fig2.show()
```
