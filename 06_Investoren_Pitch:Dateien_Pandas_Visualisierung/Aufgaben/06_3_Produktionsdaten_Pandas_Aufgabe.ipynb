## Beispiele

Mit Pandas lassen sich tabellarische Daten als `DataFrame` verarbeiten:

```python
import pandas as pd

# DataFrame aus einem Dictionary erstellen
wartung = pd.DataFrame({
    "Maschine": ["Presse", "Fräse", "Fräse", "Presse", "Drehbank", "Drehbank"],
    "Stunden": [120, 85, 200, 340, 95, 150],
    "Kosten": [4500, 3200, 7800, 12000, 3600, 5800]
})

# Erste Zeilen anzeigen
wartung.head()
```

Ein DataFrame lässt sich nach einer Bedingung filtern und nach einer Spalte sortieren:

```python
# Alle Wartungen über 100 Stunden
wartung_lang = wartung[wartung["Stunden"] > 100]
print(f"{len(wartung_lang)} Wartungen über 100 Stunden")

# Nach Kosten sortieren (absteigend)
wartung_sortiert = wartung_lang.sort_values(by="Kosten", ascending=False)
wartung_sortiert
```

Mit `groupby()` lassen sich Daten nach einer Spalte gruppieren und zusammenfassen:

```python
# Durchschnittliche Kosten pro Maschine
kosten_pro_maschine = wartung.groupby("Maschine")["Kosten"].mean()
print(kosten_pro_maschine)
```
