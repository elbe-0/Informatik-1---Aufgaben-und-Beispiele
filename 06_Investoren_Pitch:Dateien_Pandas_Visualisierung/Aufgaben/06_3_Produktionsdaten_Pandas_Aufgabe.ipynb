## Beispiele

Mit Pandas lassen sich CSV-Dateien einlesen und als `DataFrame` anzeigen:

```python
import pandas as pd

# CSV-Datei einlesen
df = pd.read_csv("../Daten/autos.csv")

# Erste 5 Zeilen anzeigen
df.head()
```

Ein DataFrame lässt sich nach einer Bedingung filtern und nach einer Spalte sortieren:

```python
# Alle Fahrzeuge ab Baujahr 2020
df_neu = df[df["Baujahr"] >= 2020]
print(f"{len(df_neu)} Fahrzeuge ab Baujahr 2020")

# Nach Baujahr sortieren (absteigend)
df_sortiert = df_neu.sort_values(by="Baujahr", ascending=False)
df_sortiert.head()
```

Mit `groupby()` lassen sich Daten nach einer Spalte gruppieren und zusammenfassen:

```python
# Anzahl Fahrzeuge pro Baujahr
anzahl_pro_jahr = df.groupby("Baujahr")["Marke"].count()
print(anzahl_pro_jahr)
```
