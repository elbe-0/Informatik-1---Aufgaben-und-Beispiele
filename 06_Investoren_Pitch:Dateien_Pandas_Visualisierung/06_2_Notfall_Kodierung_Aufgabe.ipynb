## Beispiele

### In eine Datei schreiben mit `open("w")`

Der Modus `"w"` (write) erstellt eine neue Datei oder überschreibt eine bestehende.
Mit `write()` schreiben Sie Text in die Datei:

```python
# Neue Datei erstellen und Text hineinschreiben
datei = open("test_ausgabe.txt", "w")
datei.write("Erste Zeile\n")
datei.write("Zweite Zeile\n")
datei.close()

# Kontrolle: Datei wieder einlesen
datei = open("test_ausgabe.txt", "r")
print(datei.read())
datei.close()
```

### Text anhängen mit `open("a")`

Der Modus `"a"` (append) fügt Text am Ende einer bestehenden Datei hinzu,
ohne den bisherigen Inhalt zu löschen:

```python
# Text an bestehende Datei anhängen
datei = open("test_ausgabe.txt", "a")
datei.write("Dritte Zeile (angehängt)\n")
datei.close()

# Kontrolle
datei = open("test_ausgabe.txt", "r")
print(datei.read())
datei.close()
```

### Zeichen ersetzen mit `replace()`

Die String-Methode `replace()` ersetzt alle Vorkommen eines Teilstrings durch einen anderen.
Mehrere Ersetzungen können verkettet werden:

```python
text = "HALLO WELT"
kodiert = text.replace("A", "4").replace("O", "0")
print(f"Original: {text}")
print(f"Kodiert:  {kodiert}")
```

### Zeitstempel mit `datetime`

Mit `datetime` können Sie den aktuellen Zeitpunkt ermitteln und als
formatierten String in eine Datei schreiben:

```python
import datetime

jetzt = datetime.datetime.now()
zeitstempel = jetzt.strftime("%Y-%m-%d %H:%M:%S")
print(f"Zeitstempel: {zeitstempel}")
```
