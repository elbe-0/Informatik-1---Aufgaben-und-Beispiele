## Beispiele

Die Bibliothek `matplotlib` bietet verschiedene Diagrammtypen. Hier einige Grundlagen:

### Balkendiagramm mit `plt.bar()`

```python
import matplotlib.pyplot as plt

# Einfaches Balkendiagramm
abteilungen = ["Fertigung", "Logistik", "Qualität"]
mitarbeiter = [45, 22, 18]

plt.bar(abteilungen, mitarbeiter, color="steelblue")
plt.title("Mitarbeiter pro Abteilung")
plt.ylabel("Anzahl")
plt.show()
```

### Streudiagramm mit `plt.scatter()`

```python
# Streudiagramm mit Farbe und Legende
temperatur = [20, 22, 25, 28, 30, 18, 21, 26]
leistung = [80, 82, 88, 91, 85, 78, 83, 90]
schicht = ["Tag", "Tag", "Tag", "Tag", "Nacht", "Nacht", "Nacht", "Nacht"]

# Pro Gruppe separat zeichnen, damit die Legende funktioniert
for gruppe, farbe in [("Tag", "orange"), ("Nacht", "navy")]:
    t = [temperatur[i] for i in range(len(schicht)) if schicht[i] == gruppe]
    l = [leistung[i] for i in range(len(schicht)) if schicht[i] == gruppe]
    plt.scatter(t, l, c=farbe, label=gruppe)

plt.xlabel("Temperatur (°C)")
plt.ylabel("Leistung (%)")
plt.title("Temperatur vs. Leistung")
plt.legend()
plt.show()
```

### Histogramm mit `plt.hist()`

```python
# Histogramm: Verteilung von Messwerten
messwerte = [12, 15, 14, 13, 16, 15, 14, 17, 13, 12, 15, 16, 14, 15, 13]

plt.hist(messwerte, bins=6, color="seagreen", edgecolor="black")
plt.title("Verteilung der Messwerte")
plt.xlabel("Messwert")
plt.ylabel("Häufigkeit")
plt.show()
```

### Mehrere Diagramme mit `plt.subplots()`

```python
# Zwei Diagramme nebeneinander in einem Figure
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

ax1.bar(["A", "B", "C"], [10, 20, 15], color="coral")
ax1.set_title("Diagramm 1")

ax2.hist([1, 2, 2, 3, 3, 3, 4, 4, 5], bins=5, color="skyblue", edgecolor="black")
ax2.set_title("Diagramm 2")

plt.tight_layout()
plt.show()
```

### Diagramm speichern mit `plt.savefig()`

```python
# Diagramm als PNG speichern
plt.bar(["X", "Y", "Z"], [5, 8, 3], color="mediumpurple")
plt.title("Beispiel-Export")
plt.savefig("beispiel.png", dpi=150, bbox_inches="tight")
plt.show()
print("Datei 'beispiel.png' wurde gespeichert.")
```
