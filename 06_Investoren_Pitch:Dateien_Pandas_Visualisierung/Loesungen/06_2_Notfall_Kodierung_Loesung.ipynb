## Beispiele

### In eine Datei schreiben mit `open("w")`

Der Modus `"w"` (write) erstellt eine neue Datei oder überschreibt eine bestehende.
Mit `write()` schreiben Sie Text in die Datei:

```python
# Neue Datei erstellen und Text hineinschreiben
datei = open("test_ausgabe.txt", "w")
datei.write("Erste Zeile\n")
datei.write("Zweite Zeile\n")
datei.close()

# Kontrolle: Datei wieder einlesen
datei = open("test_ausgabe.txt", "r")
print(datei.read())
datei.close()
```

### Text anhängen mit `open("a")`

Der Modus `"a"` (append) fügt Text am Ende einer bestehenden Datei hinzu,
ohne den bisherigen Inhalt zu löschen:

```python
# Text an bestehende Datei anhängen
datei = open("test_ausgabe.txt", "a")
datei.write("Dritte Zeile (angehängt)\n")
datei.close()

# Kontrolle
datei = open("test_ausgabe.txt", "r")
print(datei.read())
datei.close()
```

### Zeichen ersetzen mit `replace()`

Die String-Methode `replace()` ersetzt alle Vorkommen eines Teilstrings durch einen anderen.
Mehrere Ersetzungen können verkettet werden:

```python
text = "HALLO WELT"
kodiert = text.replace("H", "#").replace("L", "1")
print(f"Original: {text}")
print(f"Kodiert:  {kodiert}")
```

### Zeitstempel mit `datetime`

Mit `datetime` können Sie den aktuellen Zeitpunkt ermitteln und als
formatierten String in eine Datei schreiben:

```python
import datetime

jetzt = datetime.datetime.now()
zeitstempel = jetzt.strftime("%Y-%m-%d %H:%M:%S")
print(f"Zeitstempel: {zeitstempel}")
```

## Aufgabe

Machen Sie die Maschinenlogs unleserlich und schreiben Sie das Ergebnis in eine neue Datei:

**(a)** Lesen Sie die Datei `Beschreibung.txt` zeilenweise ein. Ersetzen Sie in jeder Zeile
die folgenden Großbuchstaben:

| Original | Ersetzung |
|----------|-----------|
| A        | 4         |
| B        | 8         |
| E        | 3         |
| G        | 6         |
| I        | !         |
| O        | 0         |
| T        | 7         |

Geben Sie jede kodierte Zeile auf dem Bildschirm aus.

**(b)** Schreiben Sie den kodierten Text in eine neue Datei `Kodiert.txt`.
Fügen Sie als erste Zeile einen Zeitstempel im Format `"Kodiert am: YYYY-MM-DD HH:MM:SS"` ein.

**(c)** Lesen Sie die erzeugte Datei `Kodiert.txt` zur Kontrolle ein und geben Sie den Inhalt aus.

```python
import datetime

# === Aufgabe (a): Datei einlesen und Zeichen ersetzen ===

datei = open("../Daten/Beschreibung.txt", "r")

kodierte_zeilen = []

for zeile in datei:
    kodiert = (zeile
        .replace("A", "4")
        .replace("B", "8")
        .replace("E", "3")
        .replace("G", "6")
        .replace("I", "!")
        .replace("O", "0")
        .replace("T", "7"))
    kodierte_zeilen.append(kodiert)
    print(kodiert, end="")

datei.close()

# === Aufgabe (b): Kodierten Text mit Zeitstempel in neue Datei schreiben ===

jetzt = datetime.datetime.now()
zeitstempel = jetzt.strftime("%Y-%m-%d %H:%M:%S")

ausgabe = open("Kodiert.txt", "w")
ausgabe.write(f"Kodiert am: {zeitstempel}\n")

for zeile in kodierte_zeilen:
    ausgabe.write(zeile)

ausgabe.close()

# === Aufgabe (c): Kodierte Datei zur Kontrolle einlesen ===

print("\n--- Inhalt von Kodiert.txt ---")
kontrolle = open("Kodiert.txt", "r")
print(kontrolle.read())
kontrolle.close()
```
