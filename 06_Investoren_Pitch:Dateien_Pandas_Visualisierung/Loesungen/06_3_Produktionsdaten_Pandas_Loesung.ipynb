## Beispiele

Mit Pandas lassen sich CSV-Dateien einlesen und als `DataFrame` anzeigen:

```python
import pandas as pd

# CSV-Datei einlesen
df = pd.read_csv("../Daten/autos.csv")

# Erste 5 Zeilen anzeigen
df.head()
```

Ein DataFrame lässt sich nach einer Bedingung filtern und nach einer Spalte sortieren:

```python
# Alle Fahrzeuge ab Baujahr 2020
df_neu = df[df["Baujahr"] >= 2020]
print(f"{len(df_neu)} Fahrzeuge ab Baujahr 2020")

# Nach Baujahr sortieren (absteigend)
df_sortiert = df_neu.sort_values(by="Baujahr", ascending=False)
df_sortiert.head()
```

Mit `groupby()` lassen sich Daten nach einer Spalte gruppieren und zusammenfassen:

```python
# Anzahl Fahrzeuge pro Baujahr
anzahl_pro_jahr = df.groupby("Baujahr")["Marke"].count()
print(anzahl_pro_jahr)
```

## Aufgabe

Analysieren Sie die Produktions- und Verkaufsdaten der Fabrik.

### (a) Produktionsdaten filtern und sortieren

Lesen Sie `autos.csv` ein. Filtern Sie alle Fahrzeuge mit einem Kraftstoffverbrauch unter 6 l/100km. Sortieren Sie die gefilterten Daten aufsteigend nach CO2-Ausstoß.

### (b) Durchschnittsverbrauch pro Marke

Berechnen Sie den durchschnittlichen Kraftstoffverbrauch pro Marke über den gesamten Datensatz.

### (c) Verkaufsdaten analysieren

Lesen Sie `sales.csv` ein. Filtern Sie alle Verkäufe mit einer Verkaufsmenge größer als 100. Gruppieren Sie die gefilterten Daten nach Stadt und berechnen Sie den durchschnittlichen Umsatz pro Stadt.

```python
import pandas as pd

# === Aufgabe (a): Produktionsdaten filtern und sortieren ===
df_autos = pd.read_csv("../Daten/autos.csv")

# Fahrzeuge mit Verbrauch unter 6 l/100km filtern
df_sparsam = df_autos[df_autos["Kraftstoffverbrauch (l/100km)"] < 6]
print(f"{len(df_sparsam)} Fahrzeuge mit Verbrauch unter 6 l/100km")

# Nach CO2-Ausstoß aufsteigend sortieren
df_sortiert = df_sparsam.sort_values(by="CO2-Ausstoß (g/km)")
print(df_sortiert)
```

```python
# === Aufgabe (b): Durchschnittsverbrauch pro Marke ===
df_durchschnitt = df_autos.groupby("Marke")["Kraftstoffverbrauch (l/100km)"].mean()
print(df_durchschnitt)
```

```python
# === Aufgabe (c): Verkaufsdaten analysieren ===
df_sales = pd.read_csv("../Daten/sales.csv")

# Verkaufsmenge > 100 filtern
df_viel = df_sales[df_sales["Verkaufsmenge"] > 100]
print(f"{len(df_viel)} Verkäufe mit Menge > 100")

# Nach Stadt gruppieren, Durchschnittsumsatz berechnen
df_stadt = df_viel.groupby("Stadt")["Umsatz"].mean()
print(df_stadt)
```
