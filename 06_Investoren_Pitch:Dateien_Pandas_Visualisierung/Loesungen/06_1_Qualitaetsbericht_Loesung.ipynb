## Beispiele

### Eine Textdatei öffnen und lesen

Mit `open()` öffnen Sie eine Datei. Der zweite Parameter `"r"` steht für den Lesemodus.
Mit `read()` lesen Sie den gesamten Inhalt als einen einzigen String.
Am Ende muss die Datei mit `close()` wieder geschlossen werden:

```python
# Datei öffnen, lesen, schließen
datei = open("../Daten/Beschreibung.txt", "r")
inhalt = datei.read()
print(inhalt)
datei.close()
```

### Zeile für Zeile lesen mit `readlines()`

Die Methode `readlines()` gibt eine Liste zurück, in der jedes Element eine Zeile der Datei ist:

```python
datei = open("../Daten/Beschreibung.txt", "r")
zeilen = datei.readlines()
print(f"Anzahl Zeilen: {len(zeilen)}")
print(f"Erste Zeile: {zeilen[0]}")
datei.close()
```

### Zeichen zählen mit `str.count()`

Die String-Methode `count()` zählt, wie oft ein Teilstring in einem String vorkommt:

```python
text = "Maschinenbau ist spannend"
print(f"Anzahl 'a': {text.count('a')}")
print(f"Anzahl 'A': {text.count('A')}")
print(f"Gesamt a/A: {text.count('a') + text.count('A')}")
```

### Ein einfaches Balkendiagramm mit `matplotlib`

Mit `matplotlib.pyplot` können Sie Daten als Diagramm darstellen.
`plt.bar()` erzeugt ein Balkendiagramm aus Kategorien und Werten:

```python
import matplotlib.pyplot as plt

kategorien = ["Montag", "Dienstag", "Mittwoch"]
werte = [12, 19, 8]

plt.bar(kategorien, werte)
plt.ylabel("Stückzahl")
plt.title("Produktion pro Tag")
plt.show()
```

## Aufgabe

Analysieren Sie den Qualitätsbericht `Beschreibung.txt`:

**(a)** Lesen Sie die Datei `Beschreibung.txt` zeilenweise ein. Ermitteln Sie für jede Zeile,
wie oft jeder Vokal (a, e, i, o, u - jeweils Groß- und Kleinbuchstaben zusammengezählt) vorkommt.
Geben Sie das Ergebnis pro Zeile aus.

**(b)** Ermitteln Sie, wie oft jeder Vokal im gesamten Text vorkommt.
Stellen Sie das Ergebnis als Balkendiagramm mit `matplotlib` dar.

**(c)** Erweitern Sie das Diagramm um einen aussagekräftigen Titel und eine Achsenbeschriftung.

```python
import matplotlib.pyplot as plt

# === Aufgabe (a): Vokale pro Zeile zählen und ausgeben ===

datei = open("../Daten/Beschreibung.txt", "r")

# Gesamtzähler für Aufgabe (b)
a_gesamt = 0
e_gesamt = 0
i_gesamt = 0
o_gesamt = 0
u_gesamt = 0

zeilen_nr = 1

for zeile in datei:
    a_count = zeile.count("a") + zeile.count("A")
    e_count = zeile.count("e") + zeile.count("E")
    i_count = zeile.count("i") + zeile.count("I")
    o_count = zeile.count("o") + zeile.count("O")
    u_count = zeile.count("u") + zeile.count("U")

    print(f"Zeile {zeilen_nr}: a={a_count}, e={e_count}, i={i_count}, o={o_count}, u={u_count}")

    # Gesamtzähler aktualisieren
    a_gesamt += a_count
    e_gesamt += e_count
    i_gesamt += i_count
    o_gesamt += o_count
    u_gesamt += u_count

    zeilen_nr += 1

datei.close()

# === Aufgabe (b): Gesamtergebnis als Balkendiagramm ===

vokale = ["a", "e", "i", "o", "u"]
anzahlen = [a_gesamt, e_gesamt, i_gesamt, o_gesamt, u_gesamt]

plt.bar(vokale, anzahlen)

# === Aufgabe (c): Titel und Achsenbeschriftung ===

plt.title("Vokalverteilung im Qualitätsbericht")
plt.xlabel("Vokal")
plt.ylabel("Anzahl")
plt.show()
```
